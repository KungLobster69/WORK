```python
import os
import cv2
from tqdm import tqdm

width_list = []
height_list = []

p1 = 'M1/'
for p2 in tqdm(os.listdir(p1)):
    p3 = os.path.join(p1, p2)
    img = cv2.imread(p3, cv2.IMREAD_COLOR)  # Read image in color mode
    if img is None:
        print(f"Error: Unable to load image at {p3}")
        continue  # Skip to the next image
    w, h = img.shape[:2]  # Get width and height
    width_list.append(w)
    height_list.append(h)

print("Widths:", width_list)
print("Heights:", height_list)
```

```python
def get_padding_image(img):
    width_threshold = 200
    height_threshold = 200
    pix_default = 255

    height= img.shape[0]
    width = img.shape[1]
    arr = np.zeros( (width_threshold,height_threshold, 3), dtype=np.uint8 )


    i1 = (height_threshold-height)//2
    i2 = (height_threshold-height)//2+height
    j1 = (width_threshold-width)//2
    j2 = (width_threshold-width)//2 + width
    for i in range(height_threshold):
        for j in range(width_threshold):
            if i>=i1 and i<=i2 and j>=j1 and j<=j2:
                for k in range(3):
                    if int(i-i1)<img.shape[0] and int(j-j1)<img.shape[1]:
                        arr[i,j,k] = img[ int(i-i1),  int(j-j1),   k]
                    else:
                        arr[i,j,k] = pix_default
            else:
                arr[i,j,0] = pix_default
                arr[i,j,1] = pix_default
                arr[i,j,2] = pix_default
    return arr
```

```python
import cv2,os
import numpy as np
from tqdm import *
import os
import cv2
from tqdm import *


p1 = 'O1/'
for p2 in tqdm(os.listdir(p1)):
    p3 = os.path.join(p1,p2)
    img = cv2.imread(p3, cv2.IMREAD_COLOR)# IMREAD_GRAYSCALE, IMREAD_COLOdim = (width, height)
    if img is None:
        print(f"Error: Unable to load image at {p3}")
        continue  # Skip to the next image
    # padding image
    pad_img = get_padding_image(img)
    filename_dst = p3.replace("O1","O2")
    cv2.imwrite( filename_dst, pad_img )
```
