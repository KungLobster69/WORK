```python
import cv2

f1 = '/home/ubuntu/malaria/images/efb85f8e-c780-4a19-b6ce-555a154b0d10.jpg'
img = cv2.imread(f1,cv2.IMREAD_COLOR)
img_hsv = cv2.cvtColor(img, cv2.COLOR_RGB2HSV)
channels = cv2.split(img_hsv)
clahe = cv2.createCLAHE(clipLimit = 4, tileGridSize = (20,20))
clahe.apply(channels[2],channels[2])
cv2.merge(channels, img_hsv)
img_histeq = cv2.cvtColor(img_hsv, cv2.COLOR_HSV2RGB, )
```

```python
# resize 
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import matplotlib

matplotlib.rcParams['figure.figsize'] = (16.0, 13.0)

plt.subplot(1,2,1)
plt.title('img_original')
plt.imshow(img)
plt.subplot(1,2,2)
plt.title('img_histeq')
plt.imshow(img_histeq)

plt.xticks([]), plt.yticks([])  # to hide tick values on X and Y axis
plt.show()
```

```python
# 1. histogram of non-process and processed

import os
import cv2
import numpy

def get_hist_dist(f):
    r_red =[0]*26
    r_green = [0]*26
    r_blue =[0]*26
    
   
    for i in range(img_histeq.shape[0]):
        for j in range(img_histeq.shape[1]):
            b,g,r = img_histeq[i][j]
            img[i][j] = (int(b),int(g),int(r))
            b,g,r = img_histeq[i][j]
            idx = int(b/10)
            r_blue[idx]+=1
            idx = int(g/10)
            r_green[idx]+=1
            idx = int(r/10)
            r_red[idx]+=1
            
    return r_blue,r_green,r_red

get_hist_dist(f)
```
