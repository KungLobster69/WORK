```python
import cv2
import joblib as jl

image_to_cells_true = jl.load('image_to_cells_labelme_vivax.jl')
image_to_cells_pred = jl.load('image_to_cells_pred.jl')

img = cv2.imread('vivax/img/1703121298-0002-S.jpg', cv2.IMREAD_COLOR)# IMREAD_GRAYSCALE, IMREAD_COLOR

for c in image_to_cells_true['1703121298-0002-S.jpg']:
    c1 = (int(c[0]),int(c[1]))
    c2 = (int(c[2]),int(c[3]))
    cv2.rectangle(img, c1, c2, color=(255,0,0), thickness=2)


for c in image_to_cells_pred['1703121298-0002-S.jpg']:
    c1 = (int(c[0]),int(c[1]))
    c2 = (int(c[2]),int(c[3]))
    cv2.rectangle(img, c1, c2, color=(0,0,255), thickness=2)


img2 = img[:,:,::-1]
```

```python
%matplotlib inline

from matplotlib import pyplot as plt

plt.imshow(img2)
plt.xticks([]), plt.yticks([])  # to hide tick values on X and Y axis
plt.show()
```

### test json if has error

```python
import json

with open('test_vivax.json', 'r') as fr:
    jd = json.load(fr)
```

```python
img11 = cv2.imread('1703121298-0002-S.jpg', cv2.IMREAD_COLOR)# IMREAD_GRAYSCALE, IMREAD_COLOR

for img in jd:
    pathname = img['image']['pathname']
    if '1703121298-0001-G_R.jpg' in pathname:
        for bbox in img['objects']:
            r1,c1 = bbox['bounding_box']['minimum']['r'],bbox['bounding_box']['minimum']['c']
            r2,c2 = bbox['bounding_box']['maximum']['r'],bbox['bounding_box']['maximum']['c']
            c1 = (r1,c1)
            c2 = (r2,c2)
            cv2.rectangle(img11, c1, c2, color=(255,0,0), thickness=2)
            
img22 = img11[:,:,::-1]
```

```python
%matplotlib inline

from matplotlib import pyplot as plt

plt.imshow(img22)
plt.xticks([]), plt.yticks([])  # to hide tick values on X and Y axis
plt.show()
```
