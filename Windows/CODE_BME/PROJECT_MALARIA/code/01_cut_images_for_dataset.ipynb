# 1. check cell position

## 1.1 load data

```python
import json

json_train = '../malaria_small/training.json'
with open(json_train, 'r') as fr:
    json_data_train = json.load(fr)
    
json_test = '../malaria_small/test.json'
with open(json_test, 'r') as fr:
    json_data_test = json.load(fr)

    
print('json_data_train={0}, json_data_test={1}'.format(len(json_data_train), len(json_data_test)))    
```

## 1.2 find one image for testing bounding box position meanings

```python
# find one image for testing bounding box position meanings

img = 'eeac81a5-9202-4965-b7de-35b12910cdf6.png'
pic1 = '../malaria_small/images/{0}'.format(img)
json1 = None
for d in json_data_train:
    if img in d['image']['pathname']:
        json1 = d
```

```python
json1
```

## 1.3 display selected image

```python
%matplotlib inline
from matplotlib import pyplot as plt
import numpy as np
import cv2
# display image

img = cv2.imread(pic1, cv2.IMREAD_COLOR)# IMREAD_GRAYSCALE, IMREAD_COLOR

img2 = img[:,:,::-1]

plt.imshow(img2)
plt.xticks([]), plt.yticks([])  # to hide tick values on X and Y axis
plt.show()
```

```python
img.shape
```

## 1.4 cut one cell for testing position

```python
'''
 'objects': [{'bounding_box': {'minimum': {'r': 1020, 'c': 914},
    'maximum': {'r': 1125, 'c': 1011}},
   'category': 'red blood cell'},
'''

subimg = img[1020:1125, 914:1011]
subimg2 = subimg[:,:,::-1]

plt.imshow(subimg2)
plt.xticks([]), plt.yticks([])  # to hide tick values on X and Y axis
plt.show()
```

```python
def display_sub_image(subimg):
    subimg2 = subimg[:,:,::-1]
    plt.imshow(subimg2)
    plt.xticks([]), plt.yticks([])  # to hide tick values on X and Y axis
    plt.show()
```

## 1.5 expand to all cell

```python
i = 3

x1 = json1['objects'][i]['bounding_box']['minimum']['r']
x2 = json1['objects'][i]['bounding_box']['maximum']['r']
y1 = json1['objects'][i]['bounding_box']['minimum']['c']
y2 = json1['objects'][i]['bounding_box']['maximum']['c']
label = json1['objects'][i]['category']

subimg = img[x1:x2, y1:y2]
display_sub_image(subimg)
print(label)
```
