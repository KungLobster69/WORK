```python
import pyvista as pv
import numpy as np

# สร้างจุด (Points) ของ Skeleton
points = np.array([
    [0, 0, 0],  # จุดกลาง (origin)
    [1, 0, 0],  # จุดขยับไปตามแกน x
    [0, 1, 0],  # จุดขยับไปตามแกน y
    [0, 0, 1],  # จุดขยับไปตามแกน z
    [1, 1, 1]   # จุดแบบ diagonal
])

# สร้างเส้น (Lines) ที่เชื่อมต่อจุด
lines = np.array([
    [2, 0, 1],  # เส้นจากจุด 0 ไปจุด 1
    [2, 0, 2],  # เส้นจากจุด 0 ไปจุด 2
    [2, 0, 3],  # เส้นจากจุด 0 ไปจุด 3
    [2, 1, 4],  # เส้นจากจุด 1 ไปจุด 4
    [2, 2, 4]   # เส้นจากจุด 2 ไปจุด 4
])

# สร้าง PyVista PolyData
skeleton = pv.PolyData()
skeleton.points = points
skeleton.lines = lines

# แสดง Skeleton ในตัว View
plotter = pv.Plotter()
plotter.add_mesh(skeleton, color="blue", line_width=2)
plotter.show()
```
