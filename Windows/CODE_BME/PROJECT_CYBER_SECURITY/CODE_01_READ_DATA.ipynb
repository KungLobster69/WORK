```python
import os
import glob
import random

# กำหนด path ทั้งสอง
benign_path = "/Users/kunglobster/Documents/Cyber_Security_DATA_SET/Benign"
malware_path = "/Users/kunglobster/Documents/Cyber_Security_DATA_SET/Malware"

# ใช้ glob เพื่อดึงไฟล์ที่มีรูปแบบ output<number>.txt ในแต่ละ path
benign_files = glob.glob(os.path.join(benign_path, "output*.txt"))
malware_files = glob.glob(os.path.join(malware_path, "output*.txt"))

# สร้างลิสต์เพื่อเก็บเนื้อหาของไฟล์ Benign และ Malware แยกกัน
benign_contents = []
malware_contents = []

# วนลูปเพื่อเปิดและอ่านเนื้อหาของแต่ละไฟล์ใน Benign
for file_path in benign_files:
    with open(file_path, "r", encoding="utf-8") as file:
        content = file.read()
        benign_contents.append(content)

# วนลูปเพื่อเปิดและอ่านเนื้อหาของแต่ละไฟล์ใน Malware
for file_path in malware_files:
    with open(file_path, "r", encoding="utf-8") as file:
        content = file.read()
        malware_contents.append(content)

# สุ่มข้อมูลใน benign_contents สำหรับ benign_train (1000 ตัวอย่าง) และ benign_test (ข้อมูลที่เหลือ)
benign_indices = random.sample(range(len(benign_contents)), 1000)
benign_train = [benign_contents[i] for i in benign_indices]
benign_test = [content for idx, content in enumerate(benign_contents) if idx not in benign_indices]

# สุ่มข้อมูลใน malware_contents สำหรับ malware_train (10000 ตัวอย่าง) และ malware_test (ข้อมูลที่เหลือ)
malware_indices = random.sample(range(len(malware_contents)), 10000)
malware_train = [malware_contents[i] for i in malware_indices]
malware_test = [content for idx, content in enumerate(malware_contents) if idx not in malware_indices]

# # แสดงจำนวนของแต่ละชุดข้อมูล
# print(f"จำนวนข้อมูลใน benign_train: {len(benign_train)}")
# print(f"จำนวนข้อมูลใน benign_test: {len(benign_test)}")
# print(f"จำนวนข้อมูลใน malware_train: {len(malware_train)}")
# print(f"จำนวนข้อมูลใน malware_test: {len(malware_test)}")

# แสดงดัชนีที่สุ่มมาและนับจำนวนดัชนี
print(f"ดัชนีที่สุ่มมาใน benign_train ({len(benign_indices)}): {benign_indices}")
print(f"ดัชนีที่สุ่มมาใน malware_train ({len(malware_indices)}): {malware_indices}")
```
