```python
# -*- coding: utf-8 -*-
# เฉลยแบบฝึกหัด Python พื้นฐาน 10 ข้อ
# คุณสามารถเลือก run ทีละส่วนโดยการ uncomment โค้ดในแต่ละ LAB ที่ต้องการทดสอบ

# =============================================================================
# LAB 1: โปรแกรมทักทายและคำนวณอายุ
# =============================================================================
print("--- LAB 1: โปรแกรมทักทายและคำนวณอายุ ---")
# 1. รับค่าชื่อเล่นและปีเกิดจากผู้ใช้
nickname = input("กรุณาใส่ชื่อเล่นของคุณ: ")
birth_year_str = input("กรุณาใส่ปีเกิด (พ.ศ.) ของคุณ: ")

# 2. แปลงปีเกิดเป็นตัวเลขและคำนวณอายุ
birth_year = int(birth_year_str)
current_year = 2567
age = current_year - birth_year

# 3. แสดงข้อความทักทายพร้อมบอกอายุ
print(f"สวัสดีคุณ {nickname}!")
print(f"คุณอายุ {age} ปีแล้ว")
print("-" * 20 + "\n")


# =============================================================================
# LAB 2: โปรแกรมคำนวณค่าอาหาร (หารเท่ากัน)
# =============================================================================
print("--- LAB 2: โปรแกรมคำนวณค่าอาหาร ---")
# 1. รับข้อมูลราคารวม, เปอร์เซ็นต์ทิป, และจำนวนคน
total_cost = float(input("ราคารวมของอาหารทั้งหมด: "))
tip_pct = int(input("ต้องการให้ทิปกี่เปอร์เซ็นต์ (เช่น 10, 15): "))
num_people = int(input("จำนวนคนที่หาร: "))

# 2. คำนวณทิปและยอดรวมสุทธิ
tip_amount = total_cost * (tip_pct / 100)
final_bill = total_cost + tip_amount

# 3. คำนวณค่าใช้จ่ายต่อคน
cost_per_person = final_bill / num_people

# 4. แสดงผลลัพธ์
print(f"ราคารวม: {total_cost:.2f} บาท")
print(f"ทิป {tip_pct}%: {tip_amount:.2f} บาท")
print(f"ยอดรวมสุทธิ: {final_bill:.2f} บาท")
print(f"จ่ายคนละ: {round(cost_per_person, 2):.2f} บาท")
print("-" * 20 + "\n")


# =============================================================================
# LAB 3: โปรแกรมสร้างชื่อเล่นเท่ๆ
# =============================================================================
print("--- LAB 3: โปรแกรมสร้างชื่อเล่นเท่ๆ ---")
# 1. รับชื่อจริงและเดือนเกิด
first_name = input("กรุณาใส่ชื่อจริงของคุณ: ")
birth_month = input("คุณเกิดเดือนอะไร (เช่น มกราคม): ")

# 2. สร้างชื่อเล่นตามเงื่อนไข
# - นำ 3 ตัวอักษรแรกของชื่อมาทำเป็นตัวพิมพ์ใหญ่
name_part = first_name[0:3].upper()
# - หาความยาวของชื่อเดือนเกิด
month_part = str(len(birth_month))

# 3. รวมส่วนต่างๆ เป็นชื่อเล่นใหม่
cool_nickname = name_part + month_part

# 4. แสดงผล
print(f"ชื่อเล่นสุดเท่ของคุณคือ: {cool_nickname}")
print("-" * 20 + "\n")


# =============================================================================
# LAB 4: โปรแกรมแปลงวินาทีเป็น ชั่วโมง นาที วินาที
# =============================================================================
print("--- LAB 4: โปรแกรมแปลงวินาที ---")
# 1. รับจำนวนวินาทีทั้งหมด
total_seconds = int(input("กรุณาใส่จำนวนวินาทีทั้งหมด: "))

# 2. คำนวณหา ชั่วโมง, นาที, และวินาที
# 1 ชั่วโมง = 3600 วินาที
hours = total_seconds // 3600
# หาวินาทีที่เหลือหลังจากคิดชั่วโมงไปแล้ว
remaining_seconds = total_seconds % 3600
# 1 นาที = 60 วินาที
minutes = remaining_seconds // 60
# หาวินาทีสุดท้ายที่เหลือ
seconds = remaining_seconds % 60

# 3. แสดงผล
print(f"{total_seconds} วินาที เท่ากับ {hours} ชั่วโมง {minutes} นาที {seconds} วินาที")
print("-" * 20 + "\n")


# =============================================================================
# LAB 5: โปรแกรมตรวจสอบคะแนนสอบ
# =============================================================================
print("--- LAB 5: โปรแกรมตรวจสอบคะแนนสอบ ---")
# 1. ใช้ try-except เพื่อป้องกันข้อผิดพลาด
try:
    # 2. รับคะแนนสอบ
    score_input = input("กรุณาใส่คะแนนสอบของคุณ: ")
    score = float(score_input)

    # 3. ตรวจสอบว่าคะแนนผ่านเกณฑ์หรือไม่ (>= 50)
    is_passed = score >= 50

    # 4. แสดงผลลัพธ์ (True/False)
    print(f"คะแนนของคุณคือ {score}")
    print(f"ผลการสอบ (ผ่าน/ไม่ผ่าน): {is_passed}")

except ValueError:
    # 5. กรณีผู้ใช้กรอกข้อมูลที่ไม่ใช่ตัวเลข
    print("ข้อมูลไม่ถูกต้อง! กรุณากรอกคะแนนเป็นตัวเลขเท่านั้น")
print("-" * 20 + "\n")


# =============================================================================
# LAB 6: โปรแกรมคำนวณดัชนีมวลกาย (BMI)
# =============================================================================
print("--- LAB 6: โปรแกรมคำนวณ BMI ---")
# 1. รับน้ำหนัก (กก.) และส่วนสูง (ซม.)
weight_kg = float(input("กรุณาใส่น้ำหนัก (กิโลกรัม): "))
height_cm = float(input("กรุณาใส่ส่วนสูง (เซนติเมตร): "))

# 2. แปลงส่วนสูงจากเซนติเมตรเป็นเมตร
height_m = height_cm / 100

# 3. คำนวณ BMI ตามสูตร
# สูตร: น้ำหนัก (kg) / (ส่วนสูง (m) ^ 2)
bmi = weight_kg / (height_m * height_m) # หรือ height_m ** 2

# 4. แสดงผลค่า BMI โดยปัดเศษ 1 ตำแหน่ง
print(f"ค่าดัชนีมวลกาย (BMI) ของคุณคือ: {round(bmi, 1)}")
print("-" * 20 + "\n")


# =============================================================================
# LAB 7: โปรแกรมสร้างรหัสผ่านแนะนำ
# =============================================================================
print("--- LAB 7: โปรแกรมสร้างรหัสผ่านแนะนำ ---")
# 1. รับข้อมูล
pet_name = input("ชื่อสัตว์เลี้ยงของคุณคืออะไร: ")
phone_digits = input("เลข 2 ตัวสุดท้ายของเบอร์โทรศัพท์: ")

# 2. สร้างรหัสผ่านตามเงื่อนไข
# - 4 ตัวอักษรแรกของชื่อสัตว์เลี้ยง
# - ต่อด้วยเลขท้ายเบอร์โทร
# - ต่อด้วยสัญลักษณ์พิเศษ
password = pet_name[0:4] + phone_digits + "X@!"

# 3. แสดงผล
print(f"รหัสผ่านที่แนะนำสำหรับคุณคือ: {password}")
print("-" * 20 + "\n")


# =============================================================================
# LAB 8: โปรแกรมวิเคราะห์ประโยค
# =============================================================================
print("--- LAB 8: โปรแกรมวิเคราะห์ประโยค ---")
# 1. รับประโยคจากผู้ใช้
sentence = input("ป้อนประโยคที่ต้องการวิเคราะห์: ")

# 2. นับจำนวนอักขระทั้งหมด (รวมเว้นวรรค)
char_count = len(sentence)

# 3. แยกประโยคออกเป็นคำๆ แล้วนับจำนวนคำ
words = sentence.split()
word_count = len(words)

# 4. แสดงผลการวิเคราะห์
print(f"ประโยคของคุณ: '{sentence}'")
print(f"มีจำนวนอักขระทั้งหมด: {char_count} ตัว")
print(f"มีจำนวนคำทั้งหมด: {word_count} คำ")
print("-" * 20 + "\n")


# =============================================================================
# LAB 9: โปรแกรมคำนวณส่วนลด
# =============================================================================
print("--- LAB 9: โปรแกรมคำนวณส่วนลด ---")
# 1. รับราคารวมของสินค้า
price = float(input("กรุณาใส่ราคารวมของสินค้า: "))

# 2. ตรวจสอบเงื่อนไขเพื่อรับส่วนลด
if price >= 500:
    # 3. ถ้าถึง 500 บาท ให้คำนวณส่วนลด 10%
    discount = price * 0.10
    final_price = price - discount
    print(f"คุณได้รับส่วนลด 10% เป็นเงิน {discount:.2f} บาท")
else:
    # 4. ถ้าราคาไม่ถึง 500 บาท ไม่มีส่วนลด
    final_price = price
    print("ยอดซื้อไม่ถึง 500 บาท ไม่ได้รับส่วนลด")

# 5. แสดงราคาสุทธิที่ต้องจ่าย
print(f"ราคาสุทธิที่ต้องชำระคือ: {final_price:.2f} บาท")
print("-" * 20 + "\n")


# =============================================================================
# LAB 10: โปรแกรมสลับชื่อ-นามสกุล
# =============================================================================
print("--- LAB 10: โปรแกรมสลับชื่อ-นามสกุล ---")
# 1. รับชื่อ-นามสกุลเต็ม
full_name = input("กรุณาใส่ชื่อ-นามสกุล (คั่นด้วยเว้นวรรค): ")

# 2. แยกชื่อและนามสกุลออกจากกัน
name_parts = full_name.split() # จะได้เป็น list เช่น ['สมชาย', 'ใจดี']

# 3. ดึงข้อมูลจาก list มาเก็บในตัวแปร (เพื่อความชัดเจน)
first_name = name_parts[0]
last_name = name_parts[1]

# 4. นำมาต่อกันใหม่ในรูปแบบ "นามสกุล ชื่อ"
formatted_name = f"{last_name} {first_name}"

# 5. แสดงผลลัพธ์
print(f"ชื่อในรูปแบบ 'นามสกุล ชื่อ' คือ: {formatted_name}")
print("-" * 20 + "\n")
```
