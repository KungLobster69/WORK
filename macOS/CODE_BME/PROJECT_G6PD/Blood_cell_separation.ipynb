```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

def remove_background_by_luminance(image_path, luminance_threshold=150):
    # Load the image
    image = Image.open(image_path)
    image = image.convert("RGBA")  # Convert to RGBA to support transparency
    data = np.array(image)

    # Separate color channels
    r, g, b, a = data[..., 0], data[..., 1], data[..., 2], data[..., 3]

    # Calculate luminance (brightness) using a common formula
    luminance = 0.299 * r + 0.587 * g + 0.114 * b

    # Define background mask based on luminance threshold
    background_mask = luminance > luminance_threshold

    # Set alpha channel to 0 (transparent) for background areas
    data[..., 3][background_mask] = 0

    # Display the image with transparent background
    plt.imshow(data)
    plt.axis('off')  # Hide axes
    plt.show()

# Example usage
remove_background_by_luminance('/Users/kunglobster/Documents/GitHub/WORK/macOS/CODE_BME/PROJECT_G6PD/G6PD_DATA_SET/2-5.png', luminance_threshold=150)
```
