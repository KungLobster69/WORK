```python
# -*- coding: utf-8 -*-
# บรรทัดข้างบนนี้เป็นการบอก Python ว่าไฟล์นี้ใช้ตัวอักษรภาษาไทยได้

# =================================================================
# Workshop 1: การคุยกับโปรแกรม (แนะนำตัวและของที่ชอบ)
# หัวข้อนี้เราจะฝึกรับข้อมูลจากคีย์บอร์ดและแสดงผลออกทางหน้าจอ
# =================================================================
print("\n--- Workshop 1: โปรแกรมแนะนำตัวและของที่ชอบ ---")

# input() คือคำสั่งให้โปรแกรมหยุดรอรับการพิมพ์จากเรา
# พอเราพิมพ์เสร็จแล้วกด Enter ข้อความนั้นจะถูกเก็บในตัวแปรทางซ้าย
name = input("คุณชื่ออะไร?: ")
favorite_food = input("อาหารที่ชอบคืออะไร?: ")
favorite_movie = input("ภาพยนตร์เรื่องโปรดของคุณคือเรื่องอะไร?: ")
hobby = input("งานอดิเรกของคุณคืออะไร?: ")

# แสดงผลลัพธ์
print("\n--- ผลลัพธ์ Workshop 1 ---")
# f"..." หรือ f-string เป็นวิธีง่ายๆ ในการนำค่าในตัวแปรมาแทรกในข้อความ
print(f"สวัสดีครับ ผมชื่อ {name} ชอบทาน {favorite_food} เวลาว่างชอบ {hobby} และภาพยนตร์เรื่องโปรดของผมคือ {favorite_movie} ครับ!")

# พิมพ์เส้นคั่นเพื่อความสวยงาม
print("-" * 40)
# หยุดรอให้เรากด Enter ก่อนไปต่อ
input("\nกด Enter เพื่อทำ Workshop ต่อไป...")


# =================================================================
# Workshop 2: รู้จักชนิดข้อมูล (คำนวณราคาสินค้า)
# หัวข้อนี้เราจะฝึกแปลงข้อความที่รับมาให้เป็นตัวเลขเพื่อนำไปคำนวณ
# =================================================================
print("\n--- Workshop 2: โปรแกรมคำนวณราคาสินค้า ---")

# try-except เปรียบเสมือน "ลองทำดู ถ้าพังก็ไม่ต้องตกใจ"
try:
    # ข้อมูลที่รับจาก input() จะเป็น "ข้อความ" (string) เสมอ
    price_str = input("ราคาสินค้าต่อชิ้น (บาท): ")
    quantity_str = input("ต้องการซื้อกี่ชิ้น?: ")

    # เราต้องแปลง "ข้อความ" ให้เป็น "ตัวเลข" ก่อนถึงจะคำนวณได้
    # float() คือแปลงเป็นเลขทศนิยม
    item_price = float(price_str)
    # int() คือแปลงเป็นเลขจำนวนเต็ม
    item_quantity = int(quantity_str)

    # นำตัวเลขมาคูณกันตามปกติ
    total_price = item_price * item_quantity

    print("\n--- ผลลัพธ์ Workshop 2 ---")
    # :.2f ใน f-string คือการบังคับให้แสดงผลเป็นทศนิยม 2 ตำแหน่ง (เหมาะกับราคา)
    print(f"ยอดรวมที่ต้องชำระคือ {total_price:.2f} บาท")

# ถ้าเกิดข้อผิดพลาดตอนแปลงเป็นตัวเลข (เช่น พิมพ์ "abc") โปรแกรมจะมาทำงานในบล็อกนี้
except ValueError:
    print("ข้อผิดพลาด: ดูเหมือนว่าคุณจะป้อนข้อมูลที่ไม่ใช่ตัวเลข กรุณาลองใหม่ครับ")

print("-" * 40)
input("\nกด Enter เพื่อทำ Workshop ต่อไป...")


# =================================================================
# Workshop 3: การดักจับข้อผิดพลาด (โปรแกรมหารเลขที่ปลอดภัย)
# หัวข้อนี้เราจะฝึกดักจับข้อผิดพลาดที่อาจเกิดขึ้นได้หลายแบบ
# =================================================================
print("\n--- Workshop 3: โปรแกรมหารเลขที่ปลอดภัย ---")
try:
    numerator_str = input("ป้อนตัวตั้ง: ")
    denominator_str = input("ป้อนตัวหาร: ")

    numerator = float(numerator_str)
    denominator = float(denominator_str)

    # การหารด้วยศูนย์เป็นสิ่งต้องห้ามในคณิตศาสตร์ และในโปรแกรมก็จะพังเหมือนกัน
    result = numerator / denominator

    print("\n--- ผลลัพธ์ Workshop 3 ---")
    print(f"ผลหารคือ: {result}")

# ดักจับข้อผิดพลาดกรณีที่ป้อน "ข้อความ"
except ValueError:
    print("ข้อผิดพลาด: กรุณาป้อนข้อมูลเป็นตัวเลขเท่านั้น")

# ดักจับข้อผิดพลาดกรณีที่พยายาม "หารด้วยศูนย์"
except ZeroDivisionError:
    print("ข้อผิดพลาด: ตัวหารต้องไม่เป็นศูนย์นะครับ")

print("-" * 40)
input("\nกด Enter เพื่อทำ Workshop ต่อไป...")


# =================================================================
# Workshop 4: เครื่องหมายคำนวณ (โปรแกรมแปลงวินาที)
# หัวข้อนี้เราจะใช้การหารเอาส่วน และการหารเอาเศษ มาแก้ปัญหา
# =================================================================
print("\n--- Workshop 4: โปรแกรมแปลงวินาที ---")
try:
    total_seconds_str = input("กรุณาป้อนจำนวนวินาทีทั้งหมด: ")
    total_seconds = int(total_seconds_str)

    # // คือการหารปัดเศษทิ้ง เช่น 130 // 60 จะได้ 2 (นาทีเต็ม)
    minutes_total = total_seconds // 60
    # % คือการหารเอาแต่เศษ เช่น 130 % 60 จะได้ 10 (วินาทีที่เหลือ)
    seconds_remaining = total_seconds % 60

    # นำหลักการเดิมมาใช้กับนาทีเพื่อหาชั่วโมง
    hours = minutes_total // 60
    minutes_remaining = minutes_total % 60

    print("\n--- ผลลัพธ์ Workshop 4 ---")
    print(f"{total_seconds} วินาที แปลงได้เป็น:")
    print(f"{hours} ชั่วโมง, {minutes_remaining} นาที, และ {seconds_remaining} วินาที")
except ValueError:
    print("ข้อผิดพลาด: กรุณาป้อนเป็นตัวเลขจำนวนเต็มเท่านั้น")
print("-" * 40)
input("\nกด Enter เพื่อทำ Workshop ต่อไป...")


# =================================================================
# Workshop 5: จัดการข้อความ (ตรวจสอบและจัดรูปแบบ Username)
# หัวข้อนี้เราจะใช้ฟังก์ชันสำเร็จรูปสำหรับจัดการข้อความ
# =================================================================
print("\n--- Workshop 5: โปรแกรมตรวจสอบและจัดรูปแบบ Username ---")
user_input = input("กรุณาสร้าง Username ของคุณ: ")

# Method Chaining คือการเรียกใช้ฟังก์ชันต่อกันไปเรื่อยๆ ในบรรทัดเดียว
# .strip()   -> ตัดช่องว่างหน้า-หลัง
# .lower()   -> แปลงเป็นตัวพิมพ์เล็กทั้งหมด
# .replace(' ', '_') -> หาช่องว่าง แล้วแทนที่ด้วย _
formatted_username = user_input.strip().lower().replace(' ', '_')

print("\n--- ผลลัพธ์ Workshop 5 ---")
print(f"Username ที่จัดรูปแบบแล้วคือ: {formatted_username}")

# len() คือฟังก์ชันสำหรับนับความยาวของข้อความ
username_length = len(formatted_username)

# if-else คือการสร้างเงื่อนไข "ถ้าเป็นจริงให้ทำอย่างหนึ่ง ถ้าไม่จริงให้ทำอีกอย่าง"
if 5 <= username_length <= 20:
    print("สถานะ: Username นี้ใช้ได้")
else:
    print(f"สถานะ: Username นี้ใช้ไม่ได้ (ต้องมีความยาว 5-20 ตัวอักษร แต่ตอนนี้ยาว {username_length} ตัวอักษร)")
print("-" * 40)
input("\nกด Enter เพื่อทำ Workshop ต่อไป...")


# =================================================================
# Workshop 6: เขียนโค้ดให้ดี (สร้างฟังก์ชันคำนวณ)
# หัวข้อนี้เราจะฝึกสร้าง "กล่องเครื่องมือ" (ฟังก์ชัน) ของตัวเอง เพื่อให้โค้ดสะอาดและใช้ซ้ำได้
# =================================================================
print("\n--- Workshop 6: โปรแกรมคำนวณพื้นที่และเส้นรอบรูป ---")

# def คือการสร้าง "ฟังก์ชัน" หรือ "ชุดคำสั่ง" ที่เราออกแบบเอง
# ในวงเล็บคือ "ของที่ต้องใส่เข้ามา" (parameters) เพื่อให้ฟังก์ชันทำงานได้
def calculate_rectangle_properties(length: float, width: float) -> tuple[float, float]:
    """
    ฟังก์ชันนี้เป็นเหมือนเครื่องคิดเลขสำหรับสี่เหลี่ยม
    แค่ใส่ความยาวกับความกว้างเข้าไป มันจะคืนค่าพื้นที่และเส้นรอบรูปกลับมาให้
    """
    area = length * width
    perimeter = 2 * (length + width)
    # return คือการส่ง "ผลลัพธ์" ออกจากฟังก์ชัน
    return area, perimeter

try:
    length_str = input("ป้อนความยาวของสี่เหลี่ยม: ")
    width_str = input("ป้อนความกว้างของสี่เหลี่ยม: ")
    rect_length = float(length_str)
    rect_width = float(width_str)

    # เราเรียกใช้ฟังก์ชันที่สร้างไว้เหมือนเรียกเครื่องมือ
    # ฟังก์ชันนี้คืนค่า 2 ค่า เราก็สร้างตัวแปร 2 ตัวมารับได้เลย
    rect_area, rect_perimeter = calculate_rectangle_properties(rect_length, rect_width)

    print("\n--- ผลลัพธ์ Workshop 6 ---")
    print(f"สี่เหลี่ยมขนาด {rect_length}x{rect_width} มี:")
    print(f"พื้นที่ = {rect_area:.2f}")
    print(f"เส้นรอบรูป = {rect_perimeter:.2f}")
except ValueError:
    print("ข้อผิดพลาด: กรุณาป้อนความยาวและความกว้างเป็นตัวเลข")
print("-" * 40)
print("\nสิ้นสุดโปรแกรม Workshop ทั้งหมดแล้ว")
```
