```python
class Student:
    def __init__(self, name, id):
        self.name = name # เก็บชื่อ
        self.id = id     # เก็บรหัส
```

```python
s1 = Student("Alice", "6701") # ผลลัพธ์: สร้างอินสแตนซ์ s1 พร้อม name="Alice", id="6701"
s2 = Student("Bob", "6702")   # ผลลัพธ์: สร้างอินสแตนซ์ s2 พร้อม name="Bob", id="6702"

#ตรวจสอบค่าผ่าน Attribute:
print(s1.name) # Alice
print(s2.id)   # 6702
```

```python
class Robot:
    count = 0 # Class Attribute (แชร์กันทั้งคลาส)
    
    def __init__(self, name):
        self.name = name # Instance Attribute (ของใครของมัน)
        Robot.count += 1

# ทดสอบการใช้งาน
r1 = Robot("R1")
print(f"Name: {r1.name}, Count: {Robot.count}") 
# ผลลัพธ์: Name: R1, Count: 1

r2 = Robot("R2")
print(f"Name: {r2.name}, Count: {Robot.count}") 
# ผลลัพธ์: Name: R2, Count: 2

print(f"Total Robots in System: {Robot.count}")
# ผลลัพธ์: Total Robots in System: 2
```

```python
# workshop 03
class RobotPart:
    category = "Mechanical"
    def __init__(self, name):
        self.name = name
p1 = RobotPart("Motor")
print(f"Part: {p1.name}, Type: {p1.category}")

class RobotPart:
    total_count = 0
    def __init__(self, name):
        self.name = name
        RobotPart.total_count += 1
p1 = RobotPart("Sensor")
p2 = RobotPart("Gear")
print(f"Total parts in system: {RobotPart.total_count}")

p3 = RobotPart("Frame")
print(f"Is it shared? {p1.total_count == RobotPart.total_count}")
```

```python
class Robot:
    def __init__(self, name, power):
        self.name = name
        self.power = power

    def check_status(self):
        # เข้าถึงข้อมูลภายในวัตถุผ่าน self
        if self.power > 20:
            print(f"[{self.name}] Status: Operational ({self.power}%)")
        else:
            print(f"[{self.name}] Status: Low Battery! Please charge.")

# การใช้งาน
r1 = Robot("Alpha-1", 85)
r1.check_status() # ผลลัพธ์: [Alpha-1] Status: Operational (85%)

r2 = Robot("Beta-2", 10)
r2.check_status() # ผลลัพธ์: [Beta-2] Status: Low Battery! Please charge.
```

```python
class Robot:
    def __init__(self, pin):
        self.__pin_code = pin # Private Attribute: เข้าถึงจากภายนอกไม่ได้ตรงๆ

    def get_pin(self):        # Getter Method: ช่องทางอ่านค่าที่ปลอดภัย
        return self.__pin_code

r1 = Robot(5566)
# print(r1.__pin_code) # ผลลัพธ์: AttributeError
print(r1.get_pin())    # ผลลัพธ์: 5566
```
