```python
# เฉลย
def factorial(n):
    if n == 0: return 1           # Base case
    return n * factorial(n - 1)   # Recursive step
```

```python
# เฉลย
def read_error_logs(filename):
    with open(filename, 'r') as f:
        for line in f:
            if "ERROR" in line:
                yield line  # ส่งค่าออกมาทีละตัว ไม่เก็บใส่ list

# วิธีเรียกใช้
for error_line in read_error_logs("server.log"):
    print(error_line)
```
