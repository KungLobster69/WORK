```python
import tkinter as tk

# 1. เตรียมฟังก์ชัน (Event Handler) ไว้รอรับคำสั่ง
def on_click():
    # ใช้ภาษาอังกฤษเพื่อเลี่ยงปัญหา Encoding Error บน Command Prompt ของ Windows
    print("Button Clicked!")

root = tk.Tk()

# 2. สร้างปุ่ม และเชื่อมกับฟังก์ชันด้วย command
# สังเกตว่า on_click ไม่มีวงเล็บ () ต่อท้าย
btn = tk.Button(root, text="คลิกฉันสิ", command=on_click)
btn.pack()

root.mainloop()
```

```python
# เฉลย workshop 02
import tkinter as tk

# [ระดับ 2] ฟังก์ชันที่จะถูกเรียกเมื่อกดปุ่ม
def change_text():
    # ใช้งาน .config() เพื่ออัปเดตค่าของ widget
    # [ระดับ 3] เพิ่ม fg="red" ให้เปลี่ยนสีข้อความ
    lbl_status.config(text="คลิกแล้วจ้า!", fg="red")

# [ระดับ 1] สร้างหน้าต่างหลัก
root = tk.Tk()

# [ระดับ 3] กำหนดขนาด
root.geometry("200x150")

# [ระดับ 1] และ [ระดับ 2] สร้างปุ่มและผูกฟังก์ชัน
btn_action = tk.Button(root, text="คลิกฉันสิ!", command=change_text)
btn_action.pack(pady=10)

# [ระดับ 1] สร้างข้อความตั้งต้น
lbl_status = tk.Label(root, text="ยังไม่ได้คลิก", font=("Sarabun", 12))
lbl_status.pack(pady=10)

root.mainloop()
```

```python
import tkinter as tk

root = tk.Tk()

# Checkbutton
var_agree = tk.BooleanVar()
chk = tk.Checkbutton(root, text="ฉันยอมรับเงื่อนไข", variable=var_agree)
chk.pack(pady=5)

# Radiobutton
var_gender = tk.StringVar(value="M") # ตั้งค่าเริ่มต้นเป็น M
rb1 = tk.Radiobutton(root, text="ชาย", variable=var_gender, value="M")
rb2 = tk.Radiobutton(root, text="หญิง", variable=var_gender, value="F")
rb1.pack(); rb2.pack()

root.mainloop()
```

```python
import tkinter as tk
from tkinter import ttk # ต้องนำเข้า ttk สำหรับ Combobox

root = tk.Tk()

# Listbox
listbox = tk.Listbox(root, height=3)
listbox.insert(tk.END, "Python", "Java", "C++")
listbox.pack(pady=5)

# Combobox
combo = ttk.Combobox(root, values=["ตัวเลือก A", "ตัวเลือก B"])
combo.set("กรุณาเลือก...") # ข้อความเริ่มต้น
combo.pack(pady=5)

root.mainloop()
```

```python
import tkinter as tk

root = tk.Tk()
canvas = tk.Canvas(root, width=200, height=100, bg="white")
canvas.pack(pady=10)

# วาดวงกลม/วงรี และ สี่เหลี่ยม
canvas.create_oval(10, 10, 50, 50, fill="blue")
canvas.create_rectangle(60, 10, 150, 80, fill="red")

root.mainloop()
```

```python
import tkinter as tk

root = tk.Tk()

# สร้าง Frame และจัดวางใน root
top_frame = tk.Frame(root, bg="lightgray", pady=10)
top_frame.pack(fill=tk.X)

# สร้าง Label โดยให้ top_frame เป็น Parent (ไม่ใช่ root)
lbl = tk.Label(top_frame, text="ฉันอยู่ใน Frame!", bg="yellow")
lbl.pack()

root.mainloop()
```

```python
# เฉลย workshop 03
import tkinter as tk

def submit():
    # [ระดับ 3] ประมวลผลและตรวจสอบเงื่อนไข
    if var_agree.get() == True:
        print("Success! User:", var_username.get())
    else:
        print("Error: Please accept the terms!")

root = tk.Tk()
root.title("Registration Form")
root.geometry("300x200")

# [ระดับ 1] สร้างตัวแปร
var_username = tk.StringVar()
var_agree = tk.BooleanVar()

# [ระดับ 2] สร้าง Widgets
tk.Label(root, text="Username:").pack()
entry_user = tk.Entry(root, textvariable=var_username)
entry_user.pack(pady=5)

chk = tk.Checkbutton(root, text="I accept the terms", variable=var_agree)
chk.pack(pady=5)

# [ระดับ 3] สร้างปุ่ม
btn = tk.Button(root, text="Submit", command=submit)
btn.pack(pady=10)

root.mainloop()
```

```python
import tkinter as tk

root = tk.Tk()
root.geometry("200x150")

# จัดเรียงจากบนลงล่าง และให้ขยายเต็มแกน X แนวนอน
tk.Button(root, text="Top Button", bg="red", fg="white").pack(side=tk.TOP, fill=tk.X)
tk.Button(root, text="Bottom Button", bg="blue", fg="white").pack(side=tk.BOTTOM, fill=tk.X)

# จัดเรียงซ้ายขวา
tk.Button(root, text="Left", bg="green", fg="white").pack(side=tk.LEFT, fill=tk.Y)
tk.Button(root, text="Right", bg="orange", fg="black").pack(side=tk.RIGHT, fill=tk.Y)

root.mainloop()
```
