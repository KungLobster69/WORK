```python
# เฉลย workshop 02
import tkinter as tk

# [ระดับ 2] ฟังก์ชันที่จะถูกเรียกเมื่อกดปุ่ม
def change_text():
    # ใช้งาน .config() เพื่ออัปเดตค่าของ widget
    # [ระดับ 3] เพิ่ม fg="red" ให้เปลี่ยนสีข้อความ
    lbl_status.config(text="คลิกแล้วจ้า!", fg="red")

# [ระดับ 1] สร้างหน้าต่างหลัก
root = tk.Tk()

# [ระดับ 3] กำหนดขนาด
root.geometry("200x150")

# [ระดับ 1] และ [ระดับ 2] สร้างปุ่มและผูกฟังก์ชัน
btn_action = tk.Button(root, text="คลิกฉันสิ!", command=change_text)
btn_action.pack(pady=10)

# [ระดับ 1] สร้างข้อความตั้งต้น
lbl_status = tk.Label(root, text="ยังไม่ได้คลิก", font=("Sarabun", 12))
lbl_status.pack(pady=10)

root.mainloop()
```

```python
# เฉลย workshop 03
import tkinter as tk

def submit():
    # [ระดับ 3] ประมวลผลและตรวจสอบเงื่อนไข
    if var_agree.get() == True:
        print("Success! User:", var_username.get())
    else:
        print("Error: Please accept the terms!")

root = tk.Tk()
root.title("Registration Form")
root.geometry("300x200")

# [ระดับ 1] สร้างตัวแปร
var_username = tk.StringVar()
var_agree = tk.BooleanVar()

# [ระดับ 2] สร้าง Widgets
tk.Label(root, text="Username:").pack()
entry_user = tk.Entry(root, textvariable=var_username)
entry_user.pack(pady=5)

chk = tk.Checkbutton(root, text="I accept the terms", variable=var_agree)
chk.pack(pady=5)

# [ระดับ 3] สร้างปุ่ม
btn = tk.Button(root, text="Submit", command=submit)
btn.pack(pady=10)

root.mainloop()
```

```python
import tkinter as tk

root = tk.Tk()
root.geometry("400x150")
root.title("Grid Weight Example")

# ตั้งค่าน้ำหนักการขยายให้แต่ละคอลัมน์และแถว
root.columnconfigure(0, weight=0) # คอลัมน์ 0 (ซ้ายสุด) ไม่ขยายตัว
root.columnconfigure(1, weight=1) # คอลัมน์ 1 ขยายตัว 1 ส่วน
root.columnconfigure(2, weight=2) # คอลัมน์ 2 ขยายตัว 2 เท่าของคอลัมน์ 1
root.rowconfigure(0, weight=1)    # แถว 0 ขยายตัวตามแนวตั้ง

# สร้าง Widget และใส่ sticky="nsew" เพื่อให้ขยายเต็มช่อง
lbl1 = tk.Label(root, text="ไม่ขยาย (Weight=0)", bg="red", fg="white")
lbl1.grid(row=0, column=0, sticky="nsew")

lbl2 = tk.Label(root, text="ขยาย 1 ส่วน (Weight=1)", bg="green", fg="white")
lbl2.grid(row=0, column=1, sticky="nsew")

lbl3 = tk.Label(root, text="ขยาย 2 ส่วน (Weight=2)", bg="blue", fg="white")
lbl3.grid(row=0, column=2, sticky="nsew")

root.mainloop()
```

```python
import tkinter as tk

root = tk.Tk()
root.title("Calculator Layout")

# จอแสดงผล (Display)
display = tk.Entry(root, font=("Arial", 20), justify="right")
display.grid(row=0, column=0, columnspan=4, sticky="ew", padx=5, pady=5)

# รายชื่อปุ่มที่จะนำมาสร้าง
buttons = [
    '7', '8', '9', '/',
    '4', '5', '6', '*',
    '1', '2', '3', '-',
    'C', '0', '=', '+'
]

row_val = 1
col_val = 0
for button in buttons:
    btn = tk.Button(root, text=button, width=5, height=2)
    btn.grid(row=row_val, column=col_val, padx=2, pady=2)
    col_val += 1
    if col_val > 3: # เมื่อถึงคอลัมน์สุดท้าย ให้ขึ้นบรรทัดใหม่
        col_val = 0
        row_val += 1
        
root.mainloop()
```

```python
import tkinter as tk

root = tk.Tk()
root.geometry("300x200")

# 1. ส่วนหัว (Header) ใช้ Pack
header = tk.Frame(root, bg="lightblue", height=50)
header.pack(fill=tk.X, side=tk.TOP)
tk.Label(header, text="My App Login", bg="lightblue", font=("Arial", 14)).pack(pady=10)

# 2. ส่วนเนื้อหา (Content) ใช้ Frame แล้วเอา Grid ไปไว้ข้างใน
content = tk.Frame(root)
content.pack(expand=True, fill=tk.BOTH, padx=20, pady=20)

tk.Label(content, text="Username:").grid(row=0, column=0, pady=5)
tk.Entry(content).grid(row=0, column=1, pady=5)

tk.Label(content, text="Password:").grid(row=1, column=0, pady=5)
tk.Entry(content, show="*").grid(row=1, column=1, pady=5)

tk.Button(content, text="Login", width=15).grid(row=2, column=0, columnspan=2, pady=10)

root.mainloop()
```

```python
# เฉลย workshop 3
import tkinter as tk

# [ระดับ 1] สร้างหน้าต่างหลัก
root = tk.Tk()
root.title("Numpad")

# [ระดับ 1] สร้างช่องแสดงผล
entry = tk.Entry(root, font=("Arial", 20), justify="right")
entry.grid(row=0, column=0, columnspan=3, sticky="news", padx=5, pady=5)

# [ระดับ 2] สร้างและจัดวางปุ่ม 1-9
buttons = [
    ('7', 1, 0), ('8', 1, 1), ('9', 1, 2),
    ('4', 2, 0), ('5', 2, 1), ('6', 2, 2),
    ('1', 3, 0), ('2', 3, 1), ('3', 3, 2)
]
for (text, r, c) in buttons:
    tk.Button(root, text=text, width=5, height=2).grid(row=r, column=c, padx=2, pady=2)

# [ระดับ 3] สร้างปุ่ม 0 และ . พร้อมจัด Layout
tk.Button(root, text='0', width=12, height=2).grid(row=4, column=0, columnspan=2, sticky="ew", padx=2, pady=2)
tk.Button(root, text='.', width=5, height=2).grid(row=4, column=2, padx=2, pady=2)

root.mainloop()
```

```python
import tkinter as tk

class MyApp(tk.Tk):
    def __init__(self):
        super().__init__() # สร้างหน้าต่างหลัก
        self.title("My OOP GUI")
        self.geometry("300x200")
        
        # วางโครงสร้าง UI
        self.create_widgets()

    def create_widgets(self):
        # สร้าง Widget และเก็บไว้ใน self
        self.label = tk.Label(self, text="Hello OOP GUI!", font=("Arial", 14))
        self.label.pack(pady=20)
        
        self.btn = tk.Button(self, text="Click Me", command=self.on_click)
        self.btn.pack(pady=10)
        
    def on_click(self):
        self.label.config(text="Button Clicked!", fg="red")

if __name__ == "__main__":
    app = MyApp()
    app.mainloop()
```
