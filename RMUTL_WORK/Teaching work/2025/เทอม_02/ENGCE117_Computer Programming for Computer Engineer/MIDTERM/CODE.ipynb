```python
def process_data(filename):
    f = open(filename, 'r')
    content = f.read()
    items = content.split(',')
    
    total = 0
    for item in items:
        total += int(item)
        
    print(f"Total is: {total}")
    f.close()

# การเรียกใช้งาน
process_data("my_data.txt")
```

```python
def mystery(n):
    if n <= 0:
        print("Base")
        return
    
    print(f"Start {n}")
    mystery(n - 2)
    print(f"End {n}")
```

```python
def find_errors(filename):
    results = []
    with open(filename, 'r') as f:
        for line in f:
            if "ERROR" in line:
                results.append(line)
    return results
```

```python
def find_errors_gen(filename):
    with open(filename, 'r') as f:
        for line in f:
            if "ERROR" in line:
                yield line
```

```python
class Server:
    connection_count = 0  # Class Attribute

    def __init__(self, name):
        self.name = name          # Instance Attribute
        self.active_users = 0     # Instance Attribute
        Server.connection_count += 1

    def user_login(self):
        self.active_users += 1

# การสร้าง Object
s1 = Server("Web-01")
s2 = Server("DB-01")

s1.user_login()
s1.user_login()
```
