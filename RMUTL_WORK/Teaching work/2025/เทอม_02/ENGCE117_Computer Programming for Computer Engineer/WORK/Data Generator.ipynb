```python
import csv
import random
from datetime import datetime, timedelta

def generate_inventory(filename="inventory.csv", num_items=200):
    electronic_names = [
        "Gaming Laptop", "Wireless Mouse", "Mechanical Keyboard", "Monitor 24 inch",
        "Smartphone", "Tablet", "Smartwatch", "Headphones", "Power Bank", "USB Hub",
        "Webcam", "HDMI Cable", "WiFi Router", "Printer", "Micro SD Card",
        "Drawing Tablet", "Smart Plug", "Projector", "Desk Lamp", "Extension Cord",
        "External Hard Drive", "Bluetooth Speaker", "Gaming Chair", "VR Headset", "Drone"
    ]
    perishable_names = [
        "Fresh Milk", "Whole Wheat Bread", "Organic Eggs", "Cheddar Cheese", "Greek Yogurt",
        "Apple", "Banana", "Chicken Breast", "Salmon Fillet", "Beef Steak",
        "Orange Juice", "Salted Butter", "Soft Tofu", "Grapes", "Lettuce",
        "Pork Tenderloin", "Tomatoes", "Blueberries", "Avocado", "Kimchi",
        "Vanilla Ice Cream", "Frozen Peas", "Fresh Basil", "Soy Milk", "Carrots",
        "Spinach", "Strawberries", "Ground Pork", "Sausages", "Bacon"
    ]
    
    warranty_periods = ["6 Months", "1 Year", "2 Years", "3 Years", "5 Years"]
    
    with open(filename, mode='w', newline='', encoding='utf-8') as file:
        writer = csv.writer(file)
        # Header
        writer.writerow(["ID", "Type", "Name", "Price", "Stock", "Extra_Info"])
        
        for i in range(1, num_items + 1):
            p_id = 1000 + i
            p_type = random.choice(["Electronic", "Perishable"])
            
            if p_type == "Electronic":
                base_name = random.choice(electronic_names)
                name = f"{base_name} Model-{random.randint(100, 999)}"
                price = random.randint(100, 100000)
                stock = random.randint(0, 200) 
                extra = random.choice(warranty_periods)
            else:
                base_name = random.choice(perishable_names)
                weight = random.choice([100, 200, 300, 500, 1000, 2000, 5000])
                unit = "g" if "Pack" not in base_name else ""
                
                if "Pack" in base_name:
                     name = f"{base_name} ({random.randint(2, 60)})"
                elif "ml" in base_name or "Milk" in base_name or "Juice" in base_name:
                    name = f"{base_name} ({weight}ml)"
                else:
                    name = f"{base_name} ({weight}g)"

                price = random.randint(15, 2000)
                stock = random.randint(5, 100)
                # Random expiry date within next 90 days
                days_offset = random.randint(-5, 90) 
                expiry_date = datetime.now() + timedelta(days=days_offset)
                extra = expiry_date.strftime("%Y-%m-%d")
            
            writer.writerow([p_id, p_type, name, price, stock, extra])
            
    print(f"Generated {num_items} items in {filename}")

if __name__ == "__main__":
    generate_inventory()
```
