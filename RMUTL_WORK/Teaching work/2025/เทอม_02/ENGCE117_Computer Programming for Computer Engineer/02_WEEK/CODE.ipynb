```python
import copy
a = [[1, 2], [3, 4]]  # Nested List

# 1. Shallow Copy
b = a.copy()
b[0][0] = 99          # แก้ไขไส้ในของ b
print(f"Original a: {a}")
print(f"Output: {b}") 
# Output: [[99, 2], [3, 4]]  <-- a เปลี่ยนตาม! (อันตราย)

a = [[1, 2], [3, 4]]  # Nested List
# 2. Deep Copy
c = copy.deepcopy(a)
c[1][0] = 88          # แก้ไขไส้ในของ c
print(f"Original a: {a}")
print(f"Output: {c}") 
# Output: [[99, 2], [3, 4]]  <-- a ไม่เปลี่ยนแล้ว (ปลอดภัย)
```

```python
# WorkShop_03
# Setup Data
inventory = {'apple': 10, 'banana': 20}
branch2 = {'banana': 5, 'orange': 10}

# Level 1: Check Stock
def check_stock(item):
    return inventory.get(item, 0)

print(f"Check Durian: {check_stock('durian')}") 
# Output: Check Durian: 0

# Level 2: Add Stock Function
def add_stock(item, qty):
    # ดึงค่าเก่า (ถ้าไม่มีได้ 0) + ค่าใหม่ -> บันทึกกลับ
    inventory[item] = inventory.get(item, 0) + qty

add_stock('apple', 5)
print(f"After adding apple: {inventory}")
# Output: {'apple': 15, 'banana': 20}

# Level 3: Merge Logic
for item, qty in branch2.items():
    add_stock(item, qty) # Reuse function from Level 2

print(f"Final Inventory: {inventory}")
# Output: {'apple': 15, 'banana': 25, 'orange': 10}
```

```python
# โจทย์: แปลงอุณหภูมิ Celsius เป็น Fahrenheit
celsius = [0, 10, 20, 30]

# 1. แบบเดิม (Traditional For Loop)
fahrenheit = []
for c in celsius:
    val = (c * 9/5) + 32
    fahrenheit.append(val)

# 2. แบบ List Comprehension (Pythonic)
# อ่านว่า: เอา (c * 9/5) + 32 สำหรับทุกๆ c ใน celsius
fahrenheit_comp = [(c * 9/5) + 32 for c in celsius]

print(fahrenheit_comp) 
# Output: [32.0, 50.0, 68.0, 86.0]
```

```python
matrix = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

# 1. แบบเดิม (Traditional Nested Loop)
flat_list = []
for row in matrix:          # Loop นอก (แถว)
    for num in row:         # Loop ใน (ตัวเลข)
        flat_list.append(num)

# 2. แบบ Nested Comprehension
# สังเกต: ลำดับการเขียนเหมือน Loop ปกติเป๊ะๆ (ซ้าย->ขวา)
flat_comp = [num for row in matrix for num in row]

print(flat_comp)
# Output: [1, 2, 3, 4, 5, 6, 7, 8, 9]
```

```python
from collections import deque

# 1. จัดการหัวท้าย (Head & Tail)
dq = deque(['B', 'C'])
dq.append('D')        # เข้าท้าย: ['B', 'C', 'D']
dq.appendleft('A')    # เข้าหัว:  ['A', 'B', 'C', 'D'] (เร็วมาก)

first = dq.popleft()  # ออกหัว: 'A'

# 2. Rotate (หมุนวงล้อ)
nums = deque([1, 2, 3, 4, 5])
nums.rotate(1)        # หมุนขวา 1 ที -> [5, 1, 2, 3, 4]
nums.rotate(-2)       # หมุนซ้าย 2 ที -> [2, 3, 4, 5, 1]
```

```python
from collections import Counter

# 1. โยน String ใส่ (นับตัวอักษร)
text = "mississippi"
c = Counter(text)
print(c)
# Output: Counter({'i': 4, 's': 4, 'p': 2, 'm': 1})

# 2. โยน List ใส่ (นับสินค้า)
fruits = ['apple', 'banana', 'apple', 'orange']
counts = Counter(fruits)
print(counts['apple'])  # 2
print(counts['durian']) # 0 (ไม่ Error!)
```

```python
from collections import Counter

# 1. Top N (จัดอันดับ)
c = Counter("mississippi")
print(c.most_common(2)) 
# Output: [('i', 4), ('s', 4)]

# 2. Math Operations (บวก/ลบ ผลนับ)
day1 = Counter(apple=3, banana=1)
day2 = Counter(apple=2, orange=5)
total = day1 + day2
print(total) 
# Output: Counter({'orange': 5, 'apple': 5, 'banana': 1})
```

```python
from collections import defaultdict

# 1. ใช้ 'int' เป็นตัวสร้าง (ค่าเริ่มคือ 0) - เหมาะกับการนับ/บวกคะแนน
scores = defaultdict(int)
scores['Alice'] += 10   # Key 'Alice' ถูกสร้างอัตโนมัติเป็น 0 แล้วบวก 10
scores['Bob'] += 5
print(scores)           # defaultdict(<class 'int'>, {'Alice': 10, 'Bob': 5})

# 2. ใช้ 'list' เป็นตัวสร้าง (ค่าเริ่มคือ []) - เหมาะกับการจัดกลุ่ม
groups = defaultdict(list)
groups['Fruit'].append('Apple')   # Key 'Fruit' ถูกสร้างเป็น [] แล้ว append
groups['Fruit'].append('Banana')
print(groups)           # defaultdict(<class 'list'>, {'Fruit': ['Apple', 'Banana']})
```
