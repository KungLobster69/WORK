```python
class BankAccount:
    def __init__(self, name, money):
        self.name = name
        self.money = money # Public: เปิดเผยข้อมูล ใครก็แก้ได้

# 1. สร้างบัญชี "สมชาย" มีเงิน 500 บาท
user = BankAccount("สมชาย", 500)

# 2. ความหายนะเกิดขึ้นตรงนี้! (เพราะไม่มีการป้องกัน)
user.money = -1000000    # อยู่ดีๆ เงินติดลบเป็นล้าน? (ผิดตรรกะความจริง)
user.money = "รวยแล้วจ้า"  # เอาข้อความไปใส่ในตัวเลข? (ผิดชนิดข้อมูล)
```

```python
class Car:
    def __init__(self):
        self._voltage = 12  # Protected: ควรใช้แค่ภายในคลาสหรือคลาสลูก

    def _internal_check(self):
        print(f"Checking voltage: {self._voltage}V") # OK: ใช้ภายใน

c = Car()

# การใช้งานที่ "ไม่ควรทำ" (แต่ทำได้)
c._voltage = 999  # แก้ไขค่าตัวแปร Protected จากภายนอก
print(c._voltage) # เข้าถึงตัวแปร Protected โดยตรง

# ผลเสีย: ถ้าคนเขียน Class เปลี่ยนชื่อตัวแปร _voltage เป็น _power ในเวอร์ชันหน้า
# โค้ดส่วนนี้จะพังทันที (Breaking Change)
```

```python
class User:
    def __init__(self, username, password):
        self.username = username    # Public: ใครก็เข้าถึงได้
        self.__password = password  # Private: ห้ามเข้าถึงจากภายนอก (มี __)

    def check_password(self, input_pass):
        # ภายใน Class สามารถเรียกใช้ Private ได้ปกติ
        return self.__password == input_pass

u = User("Admin", "Secret123")

# 1. ลองเรียกผ่าน Method (ทำได้)
print(u.check_password("Secret123")) # Output: True

# 2. ลองเรียกตรงๆ (Error ทันที)
# print(u.__password) 
# Result: AttributeError: 'User' object has no attribute '__password'
```

```python
class Test:
    def __init__(self):
        self.public = 10       # ตัวแปร Public: ชื่อจะเหมือนเดิม
        self.__secret = 999    # ตัวแปร Private: ชื่อจะถูกเปลี่ยน (Name Mangling)

t = Test()

# ฟังก์ชัน dir() ใช้ดูชื่อตัวแปรทั้งหมดใน object
# ลองสังเกตผลลัพธ์: จะไม่เจอ '__secret' แต่จะเจอ '_Test__secret' แทน
print(dir(t))
```

```python
# กฎคือ: _ClassName__variable
print(t._Test__secret)  # Output: 999 (อ่านค่าได้เฉยเลย!)

t._Test__secret = 0     # แก้ค่าได้ด้วย (ระบบพังได้ง่ายๆ)
print(t.public)         # ผลลัพธ์: 10
```

```python
class Citizen:
    def __init__(self, id_card):
        self.__id = id_card  # Private: ห้ามแก้

    def get_id(self):
        # เปิดช่องทางให้ "ดู" ได้อย่างเดียว
        return self.__id

person = Citizen("1-1005-99999-00-1")
print(person.get_id())  # Output: 1-1005... (อ่านได้)

# person.set_id("999")  # Error! ไม่มีเมธอดนี้ (แก้ไม่ได้)
# person.__id = "999"   # Error! เข้าถึงไม่ได้
```

```python
class User:
    def __init__(self, password):
        self.__password = password # Private

    def set_password(self, new_pass):
        # เปิดช่องทางให้ "แก้" ค่าได้
        self.__password = new_pass
        print("เปลี่ยนรหัสผ่านสำเร็จ")

u = User("1234")
u.set_password("abcd") # แก้ไขค่าผ่าน Setter
# u.__password = "xyz" # Error! แก้ตรงๆ ไม่ได้
```

```python
class Person:
    def __init__(self, age):
        self.__age = age  # Private Attribute

    def set_age(self, new_age):
        # Logic: ตรวจสอบความถูกต้องก่อนบันทึก
        if new_age > 0 and new_age < 120:
            self.__age = new_age
            print(f"บันทึกอายุ {new_age} สำเร็จ")
        else:
            print("Error: อายุไม่ถูกต้อง! (ต้องอยู่ระหว่าง 1-119)")

    def get_age(self):
        return self.__age
```

```python
# 1. สร้าง Object (อายุเริ่มต้น 20)
p = Person(20)

# 2. ลองใส่อายุติดลบ (Invalid Data)
p.set_age(-5)  
# Output: Error: อายุไม่ถูกต้อง! ... (ค่า __age ยังคงเป็น 20)

# 3. ใส่อายุที่ถูกต้อง (Valid Data)
p.set_age(25)  
# Output: บันทึกอายุ 25 สำเร็จ

print(f"อายุปัจจุบัน: {p.get_age()}")
# Output: อายุปัจจุบัน: 25
```

```python
class Rectangle:
    def __init__(self, w, h):
        self.w = w
        self.h = h

    @property
    def area(self):
        # ไม่มีตัวแปร self.area เก็บอยู่จริง
        # แต่ทุกครั้งที่เรียก .area ฟังก์ชันนี้จะทำงาน
        return self.w * self.h

r = Rectangle(4, 5)
print(f"พื้นที่เดิม: {r.area}") # Output: 20

# ลองเปลี่ยนขนาด
r.w = 10
# ค่า area จะเปลี่ยนตามทันทีโดยไม่ต้องสั่งคำนวณใหม่
print(f"พื้นที่ใหม่: {r.area}") # Output: 50 (10 * 5)
```

```python
class Circle:
    def __init__(self, radius):
        self.__radius = radius  # เก็บรัศมีเป็น Private

    # 1. สร้าง Property สำหรับอ่านค่า (Getter) แต่ไม่สร้าง Setter
    @property
    def radius(self):
        return self.__radius

    # 2. Property ที่คืนค่าคงที่ (Constant) เสมอ
    @property
    def pi(self):
        return 3.14159

c = Circle(5)

# ทดสอบการอ่าน (Read) -> ทำได้ปกติ
print(f"รัศมี: {c.radius}")  # Output: 5
print(f"ค่า PI: {c.pi}")      # Output: 3.14159

# ทดสอบการแก้ไข (Write) -> จะเกิด Error ทันที!
# c.radius = 10  # Error: AttributeError: can't set attribute
# c.pi = 3.14    # Error: AttributeError: can't set attribute
```
