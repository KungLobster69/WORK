```python
class Car:
    def __init__(self, brand, model):
        self.brand = brand
        self.model = model

    def __str__(self):
        # ‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö User: ‡∏≠‡πà‡∏≤‡∏ô‡∏á‡πà‡∏≤‡∏¢ ‡∏™‡∏ß‡∏¢‡∏á‡∏≤‡∏°
        return f"{self.brand} {self.model}"

    def __repr__(self):
        # ‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö Dev: ‡∏ä‡∏±‡∏î‡πÄ‡∏à‡∏ô ‡πÉ‡∏ä‡πâ‡∏™‡∏£‡πâ‡∏≤‡∏á Object ‡πÉ‡∏´‡∏°‡πà‡πÑ‡∏î‡πâ
        return f"Car(brand='{self.brand}', model='{self.model}')"

my_car = Car("Toyota", "Corolla")

print(str(my_car))   # ‡πÄ‡∏£‡∏µ‡∏¢‡∏Å __str__ -> Toyota Corolla
print(repr(my_car))  # ‡πÄ‡∏£‡∏µ‡∏¢‡∏Å __repr__ -> Car(brand='Toyota', model='Corolla')
```

```python
class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __str__(self):
        return f"Vector({self.x}, {self.y})"

    def __add__(self, other):
        # ‡∏ï‡∏£‡∏ß‡∏à‡∏™‡∏≠‡∏ö Type ‡πÄ‡∏û‡∏∑‡πà‡∏≠‡∏Ñ‡∏ß‡∏≤‡∏°‡∏õ‡∏•‡∏≠‡∏î‡∏†‡∏±‡∏¢
        if isinstance(other, Vector):
            return Vector(self.x + other.x, self.y + other.y)
        return NotImplemented

# ‡∏Å‡∏≤‡∏£‡πÄ‡∏£‡∏µ‡∏¢‡∏Å‡πÉ‡∏ä‡πâ‡∏á‡∏≤‡∏ô
v1 = Vector(2, 3)
v2 = Vector(5, 7)
v3 = v1 + v2
print(f"{v1} + {v2} = {v3}")

# ‡∏ú‡∏•‡∏•‡∏±‡∏û‡∏ò‡πå: Vector(2, 3) + Vector(5, 7) = Vector(7, 10)
```

```python
# ‡πÄ‡∏â‡∏•‡∏¢ workshop 01
class RPGCharacter:
    def __init__(self, name, hp, mp):
        self.name = name
        self.hp = hp
        self.mp = mp

    # Level 1: String Representation
    def __str__(self):
        return f"{self.name} (HP: {self.hp}, MP: {self.mp})"

    def __add__(self, other):
        # Level 2: Combine with another character
        if isinstance(other, RPGCharacter):
            new_name = f"{self.name}&{other.name}"
            return RPGCharacter(new_name, self.hp + other.hp, self.mp + other.mp)

        # Level 3: Heal with integer
        elif isinstance(other, int):
            return RPGCharacter(self.name, self.hp + other, self.mp)

        return NotImplemented

# Test Code
p1 = RPGCharacter("Warrior", 100, 20)
p2 = RPGCharacter("Mage", 50, 80)

print(p1)          # Warrior (HP: 100, MP: 20)
print(p1 + p2)     # Warrior&Mage (HP: 150, MP: 100)
print(p1 + 50)     # Warrior (HP: 150, MP: 20)
```

```python
class Money:
    def __init__(self, amount, currency="THB"):
        self.amount = amount
        self.currency = currency

    def __eq__(self, other):
        return self.amount == other.amount and self.currency == other.currency

    def __lt__(self, other):
        return self.amount < other.amount

# Usage
m1 = Money(100)
m2 = Money(200)
m3 = Money(100)

print(f"m1 == m3: {m1 == m3}")  # True
print(f"m1 < m2:  {m1 < m2}")   # True
print(f"m2 < m1:  {m2 < m1}")   # False
```

```python
from functools import total_ordering

@total_ordering
class Money:
    def __init__(self, amount):
        self.amount = amount

    def __eq__(self, other):
        return self.amount == other.amount

    def __lt__(self, other):
        return self.amount < other.amount

# Python ‡∏à‡∏∞‡∏™‡∏£‡πâ‡∏≤‡∏á __le__, __gt__, __ge__ ‡πÉ‡∏´‡πâ‡πÄ‡∏≠‡∏á
m1 = Money(100)
m2 = Money(200)
print(m1 <= m2) # True (‡∏ó‡∏≥‡∏á‡∏≤‡∏ô‡πÑ‡∏î‡πâ‡πÅ‡∏°‡πâ‡πÑ‡∏°‡πà‡πÑ‡∏î‡πâ‡πÄ‡∏Ç‡∏µ‡∏¢‡∏ô __le__)
```

```python
import collections

Card = collections.namedtuple('Card', ['rank', 'suit'])

class FrenchDeck:
    ranks = [str(n) for n in range(2, 11)] + list('JQKA')
    suits = 'spades diamonds clubs hearts'.split()

    def __init__(self):
        self._cards = [Card(rank, suit) for suit in self.suits
                                        for rank in self.ranks]

    def __len__(self):
        return len(self._cards)

    def __getitem__(self, position):
        return self._cards[position]

# Usage
deck = FrenchDeck()
print(f"Deck size: {len(deck)}") # ‡πÄ‡∏£‡∏µ‡∏¢‡∏Å __len__ -> 52
print(f"First card: {deck[0]}")  # ‡πÄ‡∏£‡∏µ‡∏¢‡∏Å __getitem__ -> Card(rank='2', suit='spades')
```

```python
# ‡πÄ‡∏â‡∏•‡∏¢ workshop 02
class Playlist:
    def __init__(self, songs):
        self.songs = list(songs) # Copy list to avoid reference issues

    # Level 1: Sequence Protocol
    def __len__(self):
        return len(self.songs)

    def __getitem__(self, index):
        # Level 3: Handle Slicing
        if isinstance(index, slice):
            return Playlist(self.songs[index]) # Return new Playlist Object
        return self.songs[index]

    # Level 2: Mutable & Search
    def __setitem__(self, index, value):
        self.songs[index] = value

    def __contains__(self, song):
        return song in self.songs

# Test Code
my_playlist = Playlist(["Song A", "Song B", "Song C"])

print(f"Total: {len(my_playlist)}")     # 3
print(f"First: {my_playlist[0]}")       # Song A

my_playlist[0] = "Song A (Remix)"       # Set item
print("Song B" in my_playlist)          # True

sub_playlist = my_playlist[0:2]         # Slicing
print(type(sub_playlist))               # <class '__main__.Playlist'>
print(sub_playlist.songs)               # ['Song A (Remix)', 'Song B']
```

```python
class Polynomial:
    def __init__(self, *coeffs):
        self.coeffs = coeffs

    def __call__(self, x):
        res = 0
        for index, coeff in enumerate(reversed(self.coeffs)):
            res += coeff * (x ** index)
        return res

# ‡∏™‡∏£‡πâ‡∏≤‡∏á‡∏ü‡∏±‡∏á‡∏Å‡πå‡∏ä‡∏±‡∏ô P(x) = 2x^2 + 3x + 1
p = Polynomial(2, 3, 1)

print(p(0))  # 1
print(p(2))  # 2(4) + 3(2) + 1 = 15
print(p(10)) # 2(100) + 3(10) + 1 = 231
```

```python
# 1. Decorator: ‡∏Ñ‡∏ô‡∏´‡πà‡∏≠‡∏Ç‡∏≠‡∏á‡∏Ç‡∏ß‡∏±‡∏ç
def add_bullet(func):
    def wrapper():
        return f"* {func()}" # ‡πÄ‡∏ï‡∏¥‡∏° * ‡∏Ç‡πâ‡∏≤‡∏á‡∏´‡∏ô‡πâ‡∏≤‡∏ú‡∏•‡∏•‡∏±‡∏û‡∏ò‡πå‡πÄ‡∏î‡∏¥‡∏°
    return wrapper

# 2. Function ‡πÄ‡∏î‡∏¥‡∏°: ‡∏Ç‡πâ‡∏≠‡∏Ñ‡∏ß‡∏≤‡∏°‡∏ò‡∏£‡∏£‡∏°‡∏î‡∏≤
def get_message():
    return "Hello World"

# 3. ‡∏Å‡∏≤‡∏£‡πÉ‡∏ä‡πâ‡∏á‡∏≤‡∏ô‡πÅ‡∏ö‡∏ö Manual (Reassignment)
# ‡πÄ‡∏≠‡∏≤ get_message ‡πÑ‡∏õ‡∏ú‡πà‡∏≤‡∏ô add_bullet ‡πÅ‡∏•‡πâ‡∏ß‡πÄ‡∏Å‡πá‡∏ö‡∏ó‡∏±‡∏ö‡∏ï‡∏±‡∏ß‡πÅ‡∏õ‡∏£‡πÄ‡∏î‡∏¥‡∏°
get_message = add_bullet(get_message) 

print(get_message()) 
# Output: "* Hello World" (‡∏û‡∏§‡∏ï‡∏¥‡∏Å‡∏£‡∏£‡∏°‡πÄ‡∏õ‡∏•‡∏µ‡πà‡∏¢‡∏ô‡πÑ‡∏õ‡πÅ‡∏•‡πâ‡∏ß!)
```

```python
# ‡∏õ‡∏£‡∏∞‡∏Å‡∏≤‡∏® Decorator ‡πÄ‡∏´‡∏°‡∏∑‡∏≠‡∏ô‡πÄ‡∏î‡∏¥‡∏°
def add_bullet(func):
    def wrapper():
        return f"* {func()}"
    return wrapper

# ‡πÉ‡∏ä‡πâ @add_bullet ‡πÅ‡∏õ‡∏∞‡πÑ‡∏ß‡πâ‡∏ö‡∏ô‡∏´‡∏±‡∏ß‡∏ü‡∏±‡∏á‡∏Å‡πå‡∏ä‡∏±‡∏ô‡πÄ‡∏•‡∏¢
@add_bullet 
def get_message():
    return "Hello World"

# ‡πÑ‡∏°‡πà‡∏ï‡πâ‡∏≠‡∏á‡πÄ‡∏Ç‡∏µ‡∏¢‡∏ô get_message = add_bullet(get_message) ‡∏≠‡∏µ‡∏Å‡∏ï‡πà‡∏≠‡πÑ‡∏õ
print(get_message()) 
# Output: "* Hello World" (‡∏ú‡∏•‡∏•‡∏±‡∏û‡∏ò‡πå‡πÄ‡∏´‡∏°‡∏∑‡∏≠‡∏ô‡∏Å‡∏±‡∏ô‡πÄ‡∏õ‡πä‡∏∞!)
```

```python
# ‡πÄ‡∏â‡∏•‡∏¢ workshop 03
# Level 1: Uppercase Decorator
def to_upper(func):
    def wrapper():
        original_text = func()
        return original_text.upper()
    return wrapper

# Level 2: Bracket Decorator
def add_brackets(func):
    def wrapper():
        return f"[{func()}]"
    return wrapper

# Level 3: Stacking
@add_brackets
@to_upper
def get_message():
    return "Hello Python"

print(get_message()) 
# Output: [HELLO PYTHON]
# Explanation: @to_upper ‡∏ó‡∏≥‡∏á‡∏≤‡∏ô‡∏Å‡πà‡∏≠‡∏ô (‡πÅ‡∏õ‡∏•‡∏á‡πÄ‡∏õ‡πá‡∏ô HELLO PYTHON) ‡πÅ‡∏•‡πâ‡∏ß @add_brackets ‡∏Ñ‡πà‡∏≠‡∏¢‡∏°‡∏≤‡∏Ñ‡∏£‡∏≠‡∏ö‡∏ß‡∏á‡πÄ‡∏•‡πá‡∏ö
```

```python
def log_call(func):
    # ‡∏£‡∏±‡∏ö‡∏ó‡∏∏‡∏Å‡∏≠‡∏¢‡πà‡∏≤‡∏á‡∏ó‡∏µ‡πà‡∏™‡πà‡∏á‡∏°‡∏≤ (*args = ‡πÅ‡∏ö‡∏ö‡πÄ‡∏£‡∏µ‡∏¢‡∏á, **kwargs = ‡πÅ‡∏ö‡∏ö‡∏£‡∏∞‡∏ö‡∏∏‡∏ä‡∏∑‡πà‡∏≠)
    def wrapper(*args, **kwargs):
        print(f"üì¢ Calling '{func.__name__}' with {args} {kwargs}")
        return func(*args, **kwargs) # ‡∏™‡πà‡∏á‡∏ï‡πà‡∏≠‡πÑ‡∏õ‡∏¢‡∏±‡∏á‡∏ü‡∏±‡∏á‡∏Å‡πå‡∏ä‡∏±‡∏ô‡πÄ‡∏õ‡πâ‡∏≤‡∏´‡∏°‡∏≤‡∏¢
    return wrapper

@log_call
def add(x, y):  # ‡∏£‡∏±‡∏ö 2 ‡∏ï‡∏±‡∏ß
    return x + y

@log_call
def greet(name, title="Mr."): # ‡∏£‡∏±‡∏ö 1 ‡∏ï‡∏±‡∏ß + 1 default
    print(f"Hello, {title} {name}")

# Decorator ‡∏ï‡∏±‡∏ß‡πÄ‡∏î‡∏µ‡∏¢‡∏ß ‡∏£‡∏≠‡∏á‡∏£‡∏±‡∏ö‡πÑ‡∏î‡πâ‡∏´‡∏°‡∏î!
print(add(5, 3))          # Output: üì¢ Calling 'add' with (5, 3) {} -> 8
greet("Somchai", title="Dr.") # Output: üì¢ Calling 'greet' with ('Somchai',) {'title': 'Dr.'}
```

```python
# ‚ùå ‡πÅ‡∏ö‡∏ö‡∏ú‡∏¥‡∏î: ‡∏•‡∏∑‡∏° return
def log_bad(func):
    def wrapper(*args, **kwargs):
        print("Log: Running...")
        func(*args, **kwargs) # ‡πÄ‡∏£‡∏µ‡∏¢‡∏Å‡∏ü‡∏±‡∏á‡∏Å‡πå‡∏ä‡∏±‡∏ô‡∏à‡∏£‡∏¥‡∏á ‡πÅ‡∏ï‡πà‡πÑ‡∏°‡πà‡∏™‡πà‡∏á‡∏Ñ‡πà‡∏≤‡∏Ñ‡∏∑‡∏ô!
    return wrapper

# ‚úÖ ‡πÅ‡∏ö‡∏ö‡∏ñ‡∏π‡∏Å: ‡∏£‡∏±‡∏ö‡∏Ñ‡πà‡∏≤‡πÅ‡∏•‡πâ‡∏ß‡∏™‡πà‡∏á‡∏Ñ‡∏∑‡∏ô
def log_good(func):
    def wrapper(*args, **kwargs):
        print("Log: Running...")
        result = func(*args, **kwargs) # 1. ‡πÄ‡∏Å‡πá‡∏ö‡∏Ñ‡πà‡∏≤
        return result                  # 2. ‡∏™‡πà‡∏á‡∏Ñ‡∏∑‡∏ô
    return wrapper

@log_bad
def add(a, b): return a + b

# ‡∏ú‡∏•‡∏•‡∏±‡∏û‡∏ò‡πå‡∏´‡∏≤‡∏¢!
print(f"Bad Result: {add(2, 3)}")  # Output: Bad Result: None
```

```python
def log_execution(func):
    def wrapper(*args, **kwargs):
        print(f"[LOG] Call: {func.__name__}, Args: {args} {kwargs}")
        result = func(*args, **kwargs)
        print(f"[LOG] Result: {result}")
        return result
    return wrapper

@log_execution
def divide(a, b):
    return a / b

divide(10, 2)
# Output:
# [LOG] Call: divide, Args: (10, 2) {}
# [LOG] Result: 5.0
```

```python
import time
def timer(func):
    def wrapper(*args, **kwargs):
        start = time.time()
        result = func(*args, **kwargs)
        end = time.time()
        print(f"{func.__name__} took {end-start:.4f} sec")
        return result
    return wrapper
```

```python
import functools

# ‚ùå ‡πÅ‡∏ö‡∏ö‡πÑ‡∏°‡πà‡πÉ‡∏ä‡πâ wraps
def bad_deco(func):
    def wrapper():
        func()
    return wrapper

@bad_deco
def task1():
    """Do task 1"""
    pass

print(task1.__name__) # Output: wrapper (‡∏ä‡∏∑‡πà‡∏≠‡∏à‡∏£‡∏¥‡∏á‡∏´‡∏≤‡∏¢‡πÑ‡∏õ!)
print(task1.__doc__)  # Output: None    (Docstring ‡∏´‡∏≤‡∏¢!)

# ‚úÖ ‡πÅ‡∏ö‡∏ö‡πÉ‡∏ä‡πâ wraps (Best Practice)
def good_deco(func):
    @functools.wraps(func) # <--- ‡∏Æ‡∏µ‡πÇ‡∏£‡πà‡∏Ç‡∏≠‡∏á‡πÄ‡∏£‡∏≤
    def wrapper():
        func()
    return wrapper

@good_deco
def task2():
    """Do task 2"""
    pass

print(task2.__name__) # Output: task2 (‡∏ñ‡∏π‡∏Å‡∏ï‡πâ‡∏≠‡∏á)
print(task2.__doc__)  # Output: Do task 2 (‡∏ñ‡∏π‡∏Å‡∏ï‡πâ‡∏≠‡∏á)
```

```python
import functools

def my_decorator(func):
    @functools.wraps(func)
    def wrapper(*args, **kwargs):
        # 1. Do something before
        result = func(*args, **kwargs)
        # 2. Do something after
        return result
    return wrapper
```

```python
def repeat(n):
    def decorator(func):
        def wrapper(*args, **kwargs):
            for _ in range(n):
                func(*args, **kwargs)
        return wrapper
    return decorator
```
