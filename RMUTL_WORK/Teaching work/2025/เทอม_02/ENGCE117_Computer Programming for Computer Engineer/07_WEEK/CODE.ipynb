```python
class Animal:
    def make_sound(self):
        print("... (Silence)")

class Dog(Animal):
    # Override ‡πÄ‡∏°‡∏ò‡∏≠‡∏î make_sound ‡∏Ç‡∏≠‡∏á‡πÅ‡∏°‡πà
    def make_sound(self):
        print("Woof! Woof!")

class Cat(Animal):
    # Override ‡πÄ‡∏°‡∏ò‡∏≠‡∏î make_sound ‡∏Ç‡∏≠‡∏á‡πÅ‡∏°‡πà
    def make_sound(self):
        print("Meow~")
```

```python
# ‡∏™‡∏£‡πâ‡∏≤‡∏á Zoo ‡∏ó‡∏µ‡πà‡∏°‡∏µ‡∏™‡∏±‡∏ï‡∏ß‡πå‡∏´‡∏•‡∏≤‡∏¢‡∏ä‡∏ô‡∏¥‡∏î‡∏£‡∏ß‡∏°‡∏Å‡∏±‡∏ô
zoo = [Dog(), Cat(), Dog(), Animal()]

print("--- Zoo Chorus ---")
for animal in zoo:
    animal.make_sound() 
    # ‡πÑ‡∏°‡πà‡∏ï‡πâ‡∏≠‡∏á‡πÄ‡∏ä‡πá‡∏Ñ‡∏ß‡πà‡∏≤ if type(animal) == Dog: ...
    # animal ‡πÅ‡∏ï‡πà‡∏•‡∏∞‡∏ï‡∏±‡∏ß‡∏£‡∏π‡πâ‡∏´‡∏ô‡πâ‡∏≤‡∏ó‡∏µ‡πà‡∏Ç‡∏≠‡∏á‡∏ï‡∏±‡∏ß‡πÄ‡∏≠‡∏á (Dynamic Binding)
```

```python
# Workshop Solution (‡πÄ‡∏â‡∏•‡∏¢)

class Animal:
    def make_sound(self):
        print("... (Silence)")

class Dog(Animal):
    def make_sound(self):
        print("Woof! Woof!")

class Cat(Animal):
    def make_sound(self):
        print("Meow~ Meow~")

# ‡∏Å‡∏≤‡∏£‡πÉ‡∏ä‡πâ‡∏á‡∏≤‡∏ô (Polymorphism)
band = [Dog(), Cat(), Dog(), Animal()]
print("--- The Concert Begins ---")
for member in band:
    member.make_sound()
```

```python
class Logger:
    def log(self, msg):
        print(f"[LOG] {msg}")

class TimestampLogger(Logger):
    def log(self, msg):
        import time
        # ‡πÄ‡∏£‡∏µ‡∏¢‡∏Å log ‡πÄ‡∏î‡∏¥‡∏°‡∏Ç‡∏≠‡∏á‡πÅ‡∏°‡πà ‡πÅ‡∏•‡πâ‡∏ß‡πÄ‡∏ï‡∏¥‡∏°‡πÄ‡∏ß‡∏•‡∏≤‡πÄ‡∏Ç‡πâ‡∏≤‡πÑ‡∏õ
        super().log(f"{time.ctime()} - {msg}")
```

```python
class Employee:
    def __init__(self, name, salary):
        self.name = name
        self.salary = salary

    def show_info(self):
        print(f"Employee: {self.name}, Salary: {self.salary}")
```

```python
class Manager(Employee):
    def __init__(self, name, salary, department):
        # 1. ‡πÉ‡∏´‡πâ‡πÅ‡∏°‡πà‡∏à‡∏±‡∏î‡∏Å‡∏≤‡∏£‡∏™‡πà‡∏ß‡∏ô‡∏û‡∏∑‡πâ‡∏ô‡∏ê‡∏≤‡∏ô
        super().__init__(name, salary)
        # 2. ‡∏à‡∏±‡∏î‡∏Å‡∏≤‡∏£‡∏™‡πà‡∏ß‡∏ô‡∏Ç‡∏¢‡∏≤‡∏¢
        self.department = department

    def show_info(self):
        super().show_info() # ‡πÄ‡∏£‡∏µ‡∏¢‡∏Å‡πÉ‡∏ä‡πâ method ‡πÅ‡∏°‡πà
        print(f"Department: {self.department}")

# Usage
m = Manager("Alice", 50000, "IT")
m.show_info()
```

```python
# Workshop Solution (‡πÄ‡∏â‡∏•‡∏¢)

import random

# --- Step 1: Base Class Design ---
class Character:
    def __init__(self, name, hp, power):
        self.name = name
        self.hp = hp
        self.power = power

    def attack(self, target):
        print(f"{self.name} attacks {target.name}!")
        target.take_damage(self.power)

    def take_damage(self, damage):
        self.hp -= damage
        if self.hp < 0: self.hp = 0
        print(f"\t> {self.name} took {damage} dmg. (HP left: {self.hp})")

    def is_alive(self):
        return self.hp > 0

# --- Step 2: Derived Classes ---
class Warrior(Character):
    def __init__(self, name, hp, power, defense):
        super().__init__(name, hp, power)
        self.defense = defense

    def take_damage(self, damage):
        reduced_damage = max(0, damage - self.defense)
        super().take_damage(reduced_damage)

class Mage(Character):
    def __init__(self, name, hp, power, mana):
        super().__init__(name, hp, power)
        self.mana = mana

    def attack(self, target):
        if self.mana >= 10:
            print(f"{self.name} casts Fireball! üî•")
            self.mana -= 10
            dmg = self.power + 20
        else:
            print(f"{self.name} has no mana. Hits with staff.")
            dmg = 5
        target.take_damage(dmg)

# --- Step 3: Battle Simulation ---
hero = Warrior("Arthur", 100, 15, 5)
enemy = Mage("Loki", 80, 10, 30)

print("--- BATTLE START ---")
while hero.is_alive() and enemy.is_alive():
    hero.attack(enemy)
    if enemy.is_alive():
        enemy.attack(hero)

if hero.is_alive():
    print(f"üèÜ Winner: {hero.name}")
else:
    print(f"üèÜ Winner: {enemy.name}")
```
