```python
def countdown(n):
    # ส่วนที่ 1: Base Case
    if n == 0:
        print("Blast off!")
        return  # จบการทำงานที่นี่ ไม่เรียกต่อแล้ว

    # ส่วนที่ 2: Recursive Step
    print(n)
    countdown(n - 1)  # เรียกตัวเอง แต่ลดค่า n ลง 1
```

```python
countdown(3)
```

```python
def factorial(n):
    # 1. Base Case: ต้องมีจุดหยุด ไม่งั้นติดลบไปเรื่อยๆ
    if n == 0 or n == 1:
        return 1
    
    # 2. Recursive Step:
    # คืนค่า n คูณกับ "ผลลัพธ์ของตัวก่อนหน้า"
    return n * factorial(n - 1)
```

```python
def fibonacci(n):
    # Base Cases: ต้องดัก 2 ตัวแรก
    if n == 0: return 0
    if n == 1: return 1
    
    # เรียกตัวเอง 2 ครั้ง! (แตกกิ่งก้านสาขา)
    return fibonacci(n-1) + fibonacci(n-2)
```

```python
# เฉลย Level 1

def sum_rec(n):
    # จุดหยุด
    if n == 1:
        return 1
    # เรียกต่อ
    return n + sum_rec(n - 1)

print(sum_rec(5)) # ได้ 15
```

```python
# เฉลย Level 2

def reverse_str(text):
    # จุดหยุด: ถ้าข้อความว่าง หรือเหลือตัวเดียว ไม่ต้องกลับแล้ว
    if len(text) <= 1:
        return text
    
    # เอาตัวท้ายสุด + (ส่วนที่เหลือไปกลับด้านมา)
    return text[-1] + reverse_str(text[:-1])
```

```python
# เฉลย Level 3

def power(base, exp):
    # จุดหยุด: อะไรก็ตามยกกำลัง 0 ได้ 1
    if exp == 0:
        return 1
    
    # เรียกต่อ: เอา base คูณกับก้อนที่ลดกำลังลง 1
    return base * power(base, exp - 1)
```

```python
def get_nums():
    result = []
    for i in range(1000000):
        result.append(i) # สร้างเก็บไว้ทั้งหมด
    return result
```

```python
def get_nums_gen():
    for i in range(1000000):
        yield i   # ส่งทีละตัว ไม่ต้องมี list เก็บ
```

```python
g = get_nums_gen() # สร้างตัว Generator (ยังไม่เริ่มทำงานนะ!)

print(next(g)) # เริ่มรัน -> เจอ yield 0 -> หยุด -> ได้เลข 0
print(next(g)) # ตื่นมาทำต่อ -> เจอ yield 1 -> หยุด -> ได้เลข 1
# ถ้าเรียกเกินจำนวนที่มี จะ Error ว่า StopIteration (หมดแล้วจ้า)
```

```python
for num in get_nums_gen():
    # Loop จะเรียก next() ให้เราเอง
    # และจัดการเรื่องข้อมูลหมดให้ด้วย ไม่ Error
    print(num)
```

```python
def infinite():
    n = 0
    while True:  # Loop ตลอดกาล (ปกติคอมค้างแน่ๆ)
        yield n  # แต่เพราะมี yield มันเลยหยุดรอเราทุกรอบ
        n += 1
```

```python
# เฉลย Level 1

def even_gen(limit):
    for i in range(limit + 1):
        if i % 2 == 0:
            yield i  # ส่งค่าออกไปทีละตัว

# ลองเรียกใช้
for num in even_gen(6):
    print(num) # 0, 2, 4, 6
```

```python
# เฉลย Level 2

def frange(start, stop, step):
    current = start
    while current < stop:
        yield current
        current += step

# Test
print(list(frange(0, 1.0, 0.25)))
```

```python
# เฉลย Level 3

# Gen 1: ผู้ผลิต
def gen_nums(n):
    for i in range(1, n+1):
        yield i

# Gen 2: ผู้แปรรูป
def square(numbers):
    for num in numbers:
        yield num ** 2

# การต่อท่อ
raw = gen_nums(5)       # ท่อ 1
result = square(raw)    # ท่อ 2 รับจาก ท่อ 1

print(list(result)) # [1, 4, 9, 16, 25]
```
