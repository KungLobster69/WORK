```python
def countdown(n):
    # Base Case: ตรวจสอบเงื่อนไขยุติ
    if n == 0:
        print("Blastoff!")
        return # สิ้นสุดการทำงาน

    # Recursive Step: ดำเนินการเรียกซ้ำ
    print(n)
    countdown(n - 1) # ลดขนาดของปัญหาลง
```
