```python
# ‡πÄ‡∏â‡∏•‡∏¢ Mini-Workshop 1: Basic Polymorphism
# ‡∏£‡∏ß‡∏° Level 1, 2, ‡πÅ‡∏•‡∏∞ 3

print("=== ‡πÄ‡∏£‡∏¥‡πà‡∏°‡∏ï‡πâ‡∏ô Workshop 1 ===")

# ==========================================
# üü¢ Level 1: Basic Implementation (‡∏û‡∏∑‡πâ‡∏ô‡∏ê‡∏≤‡∏ô)
# ==========================================
print("\n[Level 1 Output]")

# 1. ‡∏™‡∏£‡πâ‡∏≤‡∏á‡∏Ñ‡∏•‡∏≤‡∏™ Circle ‡πÅ‡∏•‡∏∞ Square ‡∏û‡∏£‡πâ‡∏≠‡∏° method draw
class Circle:
    def draw(self):
        print("üî¥ ‡∏ß‡∏≤‡∏î‡∏ß‡∏á‡∏Å‡∏•‡∏°")

class Square:
    def draw(self):
        print("üü¶ ‡∏ß‡∏≤‡∏î‡∏™‡∏µ‡πà‡πÄ‡∏´‡∏•‡∏µ‡πà‡∏¢‡∏°")

# 2. ‡∏™‡∏£‡πâ‡∏≤‡∏á Object ‡πÅ‡∏•‡∏∞‡πÄ‡∏£‡∏µ‡∏¢‡∏Å‡πÉ‡∏ä‡πâ‡∏á‡∏≤‡∏ô‡πÇ‡∏î‡∏¢‡∏ï‡∏£‡∏á
c = Circle()
s = Square()

c.draw()
s.draw()


# ==========================================
# üü° Level 2: Polymorphic Collection (‡∏£‡∏∞‡∏î‡∏±‡∏ö‡∏Å‡∏•‡∏≤‡∏á)
# ==========================================
print("\n[Level 2 Output]")

# 1. ‡∏ô‡∏≥‡∏Ñ‡∏•‡∏≤‡∏™‡∏à‡∏≤‡∏Å Level 1 ‡∏°‡∏≤‡πÉ‡∏ä‡πâ‡∏™‡∏£‡πâ‡∏≤‡∏á List ‡∏ó‡∏µ‡πà‡πÄ‡∏Å‡πá‡∏ö Object ‡∏ú‡∏™‡∏°‡∏Å‡∏±‡∏ô
# ‡∏™‡∏±‡∏á‡πÄ‡∏Å‡∏ï‡∏ß‡πà‡∏≤ List ‡∏ô‡∏µ‡πâ‡πÄ‡∏Å‡πá‡∏ö‡∏ó‡∏±‡πâ‡∏á Circle ‡πÅ‡∏•‡∏∞ Square ‡∏õ‡∏ô‡∏Å‡∏±‡∏ô‡πÑ‡∏î‡πâ
shapes = [Circle(), Square(), Circle(), Square()]

# 2. ‡πÉ‡∏ä‡πâ Loop ‡∏™‡∏±‡πà‡∏á‡∏á‡∏≤‡∏ô (Polymorphism)
# ‡∏ï‡∏±‡∏ß‡πÅ‡∏õ‡∏£ 'item' ‡∏à‡∏∞‡πÄ‡∏õ‡∏•‡∏µ‡πà‡∏¢‡∏ô‡∏£‡πà‡∏≤‡∏á‡πÑ‡∏õ‡πÄ‡∏£‡∏∑‡πà‡∏≠‡∏¢‡πÜ ‡∏ï‡∏≤‡∏°‡∏£‡∏≠‡∏ö‡∏Ç‡∏≠‡∏á‡∏•‡∏π‡∏õ
for item in shapes:
    item.draw()  # ‡πÄ‡∏£‡∏µ‡∏¢‡∏Å‡∏Ñ‡∏≥‡∏™‡∏±‡πà‡∏á‡πÄ‡∏î‡∏¥‡∏° ‡πÅ‡∏ï‡πà‡∏ú‡∏•‡∏•‡∏±‡∏û‡∏ò‡πå‡πÄ‡∏õ‡∏•‡∏µ‡πà‡∏¢‡∏ô‡πÑ‡∏õ‡∏ï‡∏≤‡∏° Object


# ==========================================
# üî¥ Level 3: Polymorphic Function (‡∏£‡∏∞‡∏î‡∏±‡∏ö‡∏™‡∏π‡∏á)
# ==========================================
print("\n[Level 3 Output]")

# 1. ‡∏™‡∏£‡πâ‡∏≤‡∏á‡∏Ñ‡∏•‡∏≤‡∏™‡πÉ‡∏´‡∏°‡πà Triangle (‡πÄ‡∏û‡∏∑‡πà‡∏≠‡∏à‡∏≥‡∏•‡∏≠‡∏á‡∏Å‡∏≤‡∏£‡∏Ç‡∏¢‡∏≤‡∏¢‡∏£‡∏∞‡∏ö‡∏ö‡πÉ‡∏ô‡∏≠‡∏ô‡∏≤‡∏Ñ‡∏ï)
class Triangle:
    def draw(self):
        print("üî∫ ‡∏ß‡∏≤‡∏î‡∏™‡∏≤‡∏°‡πÄ‡∏´‡∏•‡∏µ‡πà‡∏¢‡∏°")

# 2. ‡∏™‡∏£‡πâ‡∏≤‡∏á‡∏ü‡∏±‡∏á‡∏Å‡πå‡∏ä‡∏±‡∏ô‡∏Å‡∏•‡∏≤‡∏á‡∏ó‡∏µ‡πà‡∏£‡∏±‡∏ö List ‡∏Ç‡∏≠‡∏á‡∏≠‡∏∞‡πÑ‡∏£‡∏Å‡πá‡πÑ‡∏î‡πâ‡∏ó‡∏µ‡πà‡∏°‡∏µ method draw
def render_screen(shape_list):
    print(f"--- ‡∏Å‡∏≥‡∏•‡∏±‡∏á Render ‡∏†‡∏≤‡∏û‡∏à‡∏≥‡∏ô‡∏ß‡∏ô {len(shape_list)} ‡∏£‡∏π‡∏õ‡∏ó‡∏£‡∏á ---")
    for shape in shape_list:
        shape.draw() # ‡∏à‡∏∏‡∏î‡∏ô‡∏µ‡πâ‡∏Ñ‡∏∑‡∏≠ Polymorphism ‡∏ó‡∏µ‡πà‡∏ó‡∏≥‡∏á‡∏≤‡∏ô‡πÉ‡∏ô‡∏ü‡∏±‡∏á‡∏Å‡πå‡∏ä‡∏±‡∏ô

# 3. ‡πÉ‡∏ä‡πâ‡∏á‡∏≤‡∏ô‡∏à‡∏£‡∏¥‡∏á: ‡∏™‡πà‡∏á List ‡∏ó‡∏µ‡πà‡∏°‡∏µ‡∏ó‡∏±‡πâ‡∏á‡∏Ç‡∏≠‡∏á‡πÄ‡∏Å‡πà‡∏≤ (Circle, Square) ‡πÅ‡∏•‡∏∞‡∏Ç‡∏≠‡∏á‡πÉ‡∏´‡∏°‡πà (Triangle) ‡πÄ‡∏Ç‡πâ‡∏≤‡πÑ‡∏õ
my_shapes = [Circle(), Triangle(), Square()]

# ‡∏ü‡∏±‡∏á‡∏Å‡πå‡∏ä‡∏±‡∏ô render_screen ‡∏ó‡∏≥‡∏á‡∏≤‡∏ô‡πÑ‡∏î‡πâ‡πÇ‡∏î‡∏¢‡πÑ‡∏°‡πà‡∏ï‡πâ‡∏≠‡∏á‡πÅ‡∏Å‡πâ‡πÇ‡∏Ñ‡πâ‡∏î‡∏†‡∏≤‡∏¢‡πÉ‡∏ô ‡πÅ‡∏°‡πâ‡∏à‡∏∞‡∏°‡∏µ Triangle ‡πÄ‡∏û‡∏¥‡πà‡∏°‡∏°‡∏≤
render_screen(my_shapes)

print("\n=== ‡∏à‡∏ö‡∏Å‡∏≤‡∏£‡∏ó‡∏≥‡∏á‡∏≤‡∏ô ===")
```

```python
# ‡πÄ‡∏â‡∏•‡∏¢ Mini-Workshop 2: The Duck Test (Network Edition)
# ‡∏£‡∏ß‡∏° Level 1, 2, ‡πÅ‡∏•‡∏∞ 3

print("=== ‡πÄ‡∏£‡∏¥‡πà‡∏°‡∏ï‡πâ‡∏ô Workshop 2: Universal Connector ===")

# ==========================================
# üü¢ Level 1: Basic Duck Typing
# ==========================================
print("\n[Level 1 Output]")

# 1. ‡∏™‡∏£‡πâ‡∏≤‡∏á‡∏Ñ‡∏•‡∏≤‡∏™ WiFi ‡πÅ‡∏•‡∏∞ Bluetooth ‡∏ó‡∏µ‡πà‡∏°‡∏µ Interface (method) ‡πÄ‡∏´‡∏°‡∏∑‡∏≠‡∏ô‡∏Å‡∏±‡∏ô
class WiFi:
    def connect(self):
        print("üì° Connected to WiFi network")

class Bluetooth:
    def connect(self):
        print("üîµ Paired with Bluetooth device")

# 2. ‡∏™‡∏£‡πâ‡∏≤‡∏á Object (‡πÑ‡∏°‡πà‡∏ï‡πâ‡∏≠‡∏á‡∏™‡∏∑‡∏ö‡∏ó‡∏≠‡∏î‡∏à‡∏≤‡∏Å‡πÉ‡∏Ñ‡∏£)
my_wifi = WiFi()
my_bt = Bluetooth()

# 3. ‡πÄ‡∏£‡∏µ‡∏¢‡∏Å‡πÉ‡∏ä‡πâ‡∏á‡∏≤‡∏ô‡∏ï‡∏£‡∏á‡πÜ
my_wifi.connect()
my_bt.connect()


# ==========================================
# üü° Level 2: The Polymorphic Function
# ==========================================
print("\n[Level 2 Output]")

# 1. ‡∏™‡∏£‡πâ‡∏≤‡∏á‡∏ü‡∏±‡∏á‡∏Å‡πå‡∏ä‡∏±‡∏ô‡∏Å‡∏•‡∏≤‡∏á‡∏ó‡∏µ‡πà‡∏£‡∏±‡∏ö "‡∏≠‡∏∞‡πÑ‡∏£‡∏Å‡πá‡πÑ‡∏î‡πâ" ‡∏ó‡∏µ‡πà‡∏°‡∏µ .connect()
def initialize_system(device):
    print(f"--- System Initializing device: {device.__class__.__name__} ---")
    # Duck Typing: Python ‡πÑ‡∏°‡πà‡∏™‡∏ô Type ‡∏™‡∏ô‡πÅ‡∏Ñ‡πà‡∏ß‡πà‡∏≤‡∏°‡∏µ .connect() ‡πÑ‡∏´‡∏°
    device.connect()

# 2. ‡∏™‡πà‡∏á Object ‡∏ï‡πà‡∏≤‡∏á‡∏ä‡∏ô‡∏¥‡∏î‡πÄ‡∏Ç‡πâ‡∏≤‡πÑ‡∏õ‡πÉ‡∏ô‡∏ü‡∏±‡∏á‡∏Å‡πå‡∏ä‡∏±‡∏ô‡πÄ‡∏î‡∏µ‡∏¢‡∏ß‡∏Å‡∏±‡∏ô
initialize_system(my_wifi)
initialize_system(my_bt)


# ==========================================
# üî¥ Level 3: Handling Errors (EAFP)
# ==========================================
print("\n[Level 3 Output]")

# 1. ‡∏ü‡∏±‡∏á‡∏Å‡πå‡∏ä‡∏±‡∏ô‡πÅ‡∏ö‡∏ö‡∏õ‡∏•‡∏≠‡∏î‡∏†‡∏±‡∏¢ (Safe) ‡πÇ‡∏î‡∏¢‡πÉ‡∏ä‡πâ‡∏´‡∏•‡∏±‡∏Å‡∏Å‡∏≤‡∏£ EAFP
# Easier to Ask for Forgiveness than Permission
def initialize_safe(device):
    print(f"Attempting to connect: {device}")
    try:
        device.connect()
        print("‚úÖ Success!")
    except AttributeError:
        # ‡∏î‡∏±‡∏Å‡∏à‡∏±‡∏ö Error ‡πÄ‡∏°‡∏∑‡πà‡∏≠ Object ‡πÑ‡∏°‡πà‡∏°‡∏µ method connect()
        print(f"‚ùå Error: Object '{device}' ‡πÑ‡∏°‡πà‡∏™‡∏≤‡∏°‡∏≤‡∏£‡∏ñ‡πÄ‡∏ä‡∏∑‡πà‡∏≠‡∏°‡∏ï‡πà‡∏≠‡πÑ‡∏î‡πâ (‡πÑ‡∏°‡πà‡∏°‡∏µ‡∏ü‡∏±‡∏á‡∏Å‡πå‡∏ä‡∏±‡∏ô connect)")

# 2. ‡∏ó‡∏î‡∏™‡∏≠‡∏ö‡∏Å‡∏±‡∏ö‡∏Ç‡∏≠‡∏á‡∏à‡∏£‡∏¥‡∏á
initialize_safe(my_wifi)

print("-" * 20)

# 3. ‡∏ó‡∏î‡∏™‡∏≠‡∏ö‡∏Å‡∏±‡∏ö‡∏Ç‡∏≠‡∏á‡∏õ‡∏•‡∏≠‡∏° (String) ‡πÄ‡∏û‡∏∑‡πà‡∏≠‡∏î‡∏π Error Handling
fake_device = "‡∏™‡∏≤‡∏¢‡πÅ‡∏•‡∏ô‡∏õ‡∏•‡∏≠‡∏°"
initialize_safe(fake_device)

print("\n=== ‡∏à‡∏ö‡∏Å‡∏≤‡∏£‡∏ó‡∏≥‡∏á‡∏≤‡∏ô ===")
```

```python
from abc import ABC, abstractmethod

class Template(ABC):    # 1. Inherit from ABC helper class
    @abstractmethod     # 2. Decorator to mark as abstract
    def do_something(self):
        pass            # 3. Empty body
```

```python
class Incomplete(Template):
    pass # Forgot to implement do_something

obj = Incomplete() 
# TypeError: Can't instantiate abstract class Incomplete 
# with abstract method do_something
```

```python
class Complete(Template):
    def do_something(self):
        print("Done!") # Contract fulfilled

obj = Complete() # Instantiation successful
```

```python
from abc import ABC, abstractmethod

# ==========================================
# Level 1: The Strict Blueprint
# ==========================================
print("--- Level 1 ---")

# 1. ‡∏™‡∏£‡πâ‡∏≤‡∏á Abstract Class IoTDevice
class IoTDevice(ABC):
    @abstractmethod
    def turn_on(self):
        pass

    @abstractmethod
    def turn_off(self):
        pass

# 2. ‡∏™‡∏£‡πâ‡∏≤‡∏á‡∏Ñ‡∏•‡∏≤‡∏™ SmartLight ‡∏ó‡∏µ‡πà‡∏™‡∏∑‡∏ö‡∏ó‡∏≠‡∏î‡πÅ‡∏•‡∏∞ Implement ‡∏Ñ‡∏£‡∏ö
class SmartLight(IoTDevice):
    def turn_on(self):
        print("‡πÑ‡∏ü‡πÄ‡∏õ‡∏¥‡∏î")
        
    def turn_off(self):
        print("‡πÑ‡∏ü‡∏õ‡∏¥‡∏î")

# 3. ‡∏ó‡∏î‡∏™‡∏≠‡∏ö‡∏™‡∏£‡πâ‡∏≤‡∏á Object ‡πÅ‡∏•‡∏∞‡πÄ‡∏£‡∏µ‡∏¢‡∏Å‡πÉ‡∏ä‡πâ
light = SmartLight()
light.turn_on()
light.turn_off()


# ==========================================
# Level 2: Breaking the Contract
# ==========================================
print("\n--- Level 2 ---")

# 1. ‡∏™‡∏£‡πâ‡∏≤‡∏á‡∏Ñ‡∏•‡∏≤‡∏™ SmartSpeaker ‡∏ó‡∏µ‡πà Implement ‡πÑ‡∏°‡πà‡∏Ñ‡∏£‡∏ö (‡∏•‡∏∑‡∏° turn_off)
class SmartSpeaker(IoTDevice):
    def turn_on(self):
        print("‡∏•‡∏≥‡πÇ‡∏û‡∏á‡πÄ‡∏õ‡∏¥‡∏î")
    # ‡∏Ç‡∏≤‡∏î turn_off() ‡πÑ‡∏õ

# 2. ‡πÉ‡∏ä‡πâ try...except ‡∏î‡∏±‡∏Å‡∏à‡∏±‡∏ö Error ‡∏ï‡∏≠‡∏ô‡∏™‡∏£‡πâ‡∏≤‡∏á Object
try:
    speaker = SmartSpeaker()
except TypeError as e:
    print(f"Error: {e}")


# ==========================================
# Level 3: Safe Integration
# ==========================================
print("\n--- Level 3 ---")

def test_device(device):
    # ‡∏ï‡∏£‡∏ß‡∏à‡∏™‡∏≠‡∏ö‡∏ß‡πà‡∏≤‡πÄ‡∏õ‡πá‡∏ô‡∏•‡∏π‡∏Å‡∏´‡∏•‡∏≤‡∏ô‡∏Ç‡∏≠‡∏á IoTDevice ‡∏´‡∏£‡∏∑‡∏≠‡πÑ‡∏°‡πà
    if isinstance(device, IoTDevice):
        print(f"Testing {type(device).__name__}...")
        device.turn_on()
        device.turn_off()
    else:
        print("! ‡πÑ‡∏°‡πà‡πÉ‡∏ä‡πà‡∏≠‡∏∏‡∏õ‡∏Å‡∏£‡∏ì‡πå IoT ‡∏ó‡∏µ‡πà‡∏£‡∏≠‡∏á‡∏£‡∏±‡∏ö")

# ‡∏ó‡∏î‡∏™‡∏≠‡∏ö‡∏™‡πà‡∏á SmartLight (‡∏ú‡πà‡∏≤‡∏ô)
test_device(light)

# ‡∏ó‡∏î‡∏™‡∏≠‡∏ö‡∏™‡πà‡∏á String (‡πÑ‡∏°‡πà‡∏ú‡πà‡∏≤‡∏ô)
test_device("Smart Fridge (Fake)")
```
